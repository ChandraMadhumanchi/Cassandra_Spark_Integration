```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
import warnings
warnings.filterwarnings('ignore')

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

class RestaurantTurnoverPredictor:
    def __init__(self):
        self.scaler_features = StandardScaler()
        self.scaler_target = StandardScaler()  # Changed from MinMaxScaler
        self.label_encoders = {}
        self.model = None

    def load_data(self, train_path, test_path):
        """Load training and test datasets"""
        self.train_df = pd.read_csv(train_path)
        self.test_df = pd.read_csv(test_path)
        print(f"Training data shape: {self.train_df.shape}")
        print(f"Test data shape: {self.test_df.shape}")
        return self.train_df, self.test_df

    def explore_data(self):
        """Perform basic data exploration"""
        print("\n=== DATA EXPLORATION ===")
        print("\nTraining Data Info:")
        print(self.train_df.info())
        print(f"\nMissing values in training data:\n{self.train_df.isnull().sum()}")
        print(f"\nTarget variable statistics:")
        print(self.train_df['Annual Turnover'].describe())

        # Check for zeros in target
        zero_count = (self.train_df['Annual Turnover'] == 0).sum()
        print(f"Number of zero targets: {zero_count}")

        # Visualize target distribution
        plt.figure(figsize=(12, 4))

        plt.subplot(1, 2, 1)
        plt.hist(self.train_df['Annual Turnover'], bins=50, edgecolor='black', alpha=0.7)
        plt.title('Distribution of Annual Turnover')
        plt.xlabel('Annual Turnover (INR)')
        plt.ylabel('Frequency')

        plt.subplot(1, 2, 2)
        plt.hist(np.log1p(self.train_df['Annual Turnover']), bins=50, edgecolor='black', alpha=0.7)
        plt.title('Log-transformed Annual Turnover')
        plt.xlabel('Log(Annual Turnover + 1)')
        plt.ylabel('Frequency')

        plt.tight_layout()
        plt.show()

    def preprocess_data(self):
        """Comprehensive data preprocessing"""
        print("\n=== DATA PREPROCESSING ===")

        # Combine train and test for consistent preprocessing
        combined_df = pd.concat([self.train_df, self.test_df], ignore_index=True, sort=False)

        # Handle date features
        if 'Opening Day of Restaurant' in combined_df.columns:
            combined_df['Opening Day of Restaurant'] = pd.to_datetime(
                combined_df['Opening Day of Restaurant'], errors='coerce'
            )
            # Extract useful date features
            combined_df['Opening_Year'] = combined_df['Opening Day of Restaurant'].dt.year
            combined_df['Opening_Month'] = combined_df['Opening Day of Restaurant'].dt.month
            combined_df['Opening_DayOfYear'] = combined_df['Opening Day of Restaurant'].dt.dayofyear
            combined_df['Days_Since_Opening'] = (pd.Timestamp.now() - combined_df['Opening Day of Restaurant']).dt.days
            combined_df = combined_df.drop('Opening Day of Restaurant', axis=1)

        # Handle missing values
        print("Handling missing values...")

        # Numerical columns
        numerical_cols = combined_df.select_dtypes(include=[np.number]).columns
        for col in numerical_cols:
            if col != 'Annual Turnover':
                combined_df[col] = combined_df[col].fillna(combined_df[col].median())

        # Categorical columns
        categorical_cols = combined_df.select_dtypes(include=['object']).columns
        for col in categorical_cols:
            if col != 'Registration Number':
                combined_df[col] = combined_df[col].fillna(combined_df[col].mode()[0] if not combined_df[col].mode().empty else 'Unknown')

        # Encode categorical variables
        print("Encoding categorical variables...")
        for col in categorical_cols:
            if col not in ['Registration Number']:
                le = LabelEncoder()
                combined_df[col] = le.fit_transform(combined_df[col].astype(str))
                self.label_encoders[col] = le

        # Feature engineering
        print("Creating new features...")

        # Social media engagement score
        if 'Facebook Popularity Quotient' in combined_df.columns and 'Instagram Popularity Quotient' in combined_df.columns:
            combined_df['Social_Media_Score'] = (
                combined_df['Facebook Popularity Quotient'] +
                combined_df['Instagram Popularity Quotient']
            ) / 2

        # Service quality composite score
        service_cols = ['Staff Responsivness', 'Hygiene Rating', 'Food Rating', 'Service']
        available_service_cols = [col for col in service_cols if col in combined_df.columns]
        if available_service_cols:
            combined_df['Service_Quality_Score'] = combined_df[available_service_cols].mean(axis=1)

        # Ambience and entertainment score
        ambience_cols = ['Ambience', 'Lively', 'Live Music Rating', 'Comedy Gigs Rating']
        available_ambience_cols = [col for col in ambience_cols if col in combined_df.columns]
        if available_ambience_cols:
            combined_df['Entertainment_Score'] = combined_df[available_ambience_cols].mean(axis=1)

        # Infrastructure score
        infra_cols = ['Fire Audit', 'Liquor License Obtained', 'Situated in a Multi Complex',
                      'Dedicated Parking', 'Open Sitting Available']
        available_infra_cols = [col for col in infra_cols if col in combined_df.columns]
        if available_infra_cols:
            combined_df['Infrastructure_Score'] = combined_df[available_infra_cols].sum(axis=1)

        # Split back into train and test
        train_size = len(self.train_df)
        self.processed_train = combined_df[:train_size].copy()
        self.processed_test = combined_df[train_size:].copy()

        print(f"Processed training data shape: {self.processed_train.shape}")
        print(f"Processed test data shape: {self.processed_test.shape}")

        return self.processed_train, self.processed_test

    def prepare_lstm_data(self, sequence_length=5):  # Reduced sequence length
        """Prepare data for LSTM model - FIXED VERSION"""
        print(f"\n=== PREPARING LSTM DATA ===")

        # Prepare features (exclude Registration Number and Annual Turnover)
        feature_cols = [col for col in self.processed_train.columns
                       if col not in ['Registration Number', 'Annual Turnover']]

        X_train = self.processed_train[feature_cols].values
        y_train = self.processed_train['Annual Turnover'].values
        X_test = self.processed_test[feature_cols].values

        print(f"Original X_train shape: {X_train.shape}")
        print(f"Original y_train shape: {y_train.shape}")
        print(f"Target statistics before scaling:")
        print(f"Min: {y_train.min():,.2f}, Max: {y_train.max():,.2f}, Mean: {y_train.mean():,.2f}")

        # Scale features
        X_train_scaled = self.scaler_features.fit_transform(X_train)
        X_test_scaled = self.scaler_features.transform(X_test)

        # Scale target variable - FIXED: Use log transform with proper scaling
        y_train_log = np.log1p(y_train)  # log(1+x) to handle any zeros
        y_train_scaled = self.scaler_target.fit_transform(y_train_log.reshape(-1, 1)).flatten()

        print(f"Target after log transform - Min: {y_train_log.min():.2f}, Max: {y_train_log.max():.2f}")
        print(f"Target after scaling - Min: {y_train_scaled.min():.2f}, Max: {y_train_scaled.max():.2f}")

        # FIXED: Better sequence creation for non-time-series data
        # Instead of replicating data, use sliding windows with noise
        def create_sequences_robust(data, target, seq_length):
            n_samples = data.shape[0]
            n_features = data.shape[1]

            # Create more sequences by adding small random noise
            sequences = []
            targets = []

            for i in range(n_samples):
                # Create multiple variations of each sample
                for rep in range(seq_length):
                    seq = []
                    for j in range(seq_length):
                        # Add small random noise to create variation
                        noise = np.random.normal(0, 0.01, n_features)
                        sample_idx = (i + j) % n_samples
                        noisy_sample = data[sample_idx] + noise
                        seq.append(noisy_sample)

                    sequences.append(seq)
                    targets.append(target[i])

            return np.array(sequences), np.array(targets)

        # Create sequences
        X_train_seq, y_train_seq = create_sequences_robust(X_train_scaled, y_train_scaled, sequence_length)
        X_test_seq, _ = create_sequences_robust(X_test_scaled, np.zeros(len(X_test_scaled)), sequence_length)

        print(f"LSTM training sequences shape: {X_train_seq.shape}")
        print(f"LSTM training targets shape: {y_train_seq.shape}")
        print(f"LSTM test sequences shape: {X_test_seq.shape}")

        # Store original test size for prediction averaging
        self.original_test_size = len(X_test)

        return X_train_seq, y_train_seq, X_test_seq, len(feature_cols)

    def build_lstm_model(self, input_shape, n_features):
        """Build and compile LSTM model - SIMPLIFIED VERSION"""
        print(f"\n=== BUILDING LSTM MODEL ===")
        print(f"Input shape: {input_shape}")

        model = Sequential([
            # Simplified architecture
            LSTM(64, return_sequences=True, input_shape=input_shape, dropout=0.2),
            LSTM(32, return_sequences=False, dropout=0.2),

            Dense(64, activation='relu'),
            Dropout(0.3),

            Dense(32, activation='relu'),
            Dropout(0.2),

            Dense(1, activation='linear')  # CRITICAL: No activation for regression
        ])

        # Use smaller learning rate
        optimizer = Adam(learning_rate=0.0001)
        model.compile(optimizer=optimizer, loss='mse', metrics=['mae'])

        print(model.summary())
        return model

    def train_model(self, X_train, y_train, epochs=200, batch_size=16, validation_split=0.2):
        """Train the LSTM model"""
        print(f"\n=== TRAINING LSTM MODEL ===")

        # More conservative callbacks
        callbacks = [
            EarlyStopping(
                monitor='val_loss',
                patience=30,
                restore_best_weights=True,
                verbose=1
            ),
            ReduceLROnPlateau(
                monitor='val_loss',
                factor=0.5,
                patience=15,
                min_lr=1e-7,
                verbose=1
            )
        ]

        # Train model
        history = self.model.fit(
            X_train, y_train,
            epochs=epochs,
            batch_size=batch_size,
            validation_split=validation_split,
            callbacks=callbacks,
            verbose=1,
            shuffle=True
        )

        return history

    def plot_training_history(self, history):
        """Plot training history"""
        plt.figure(figsize=(15, 5))

        plt.subplot(1, 2, 1)
        plt.plot(history.history['loss'], label='Training Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.title('Model Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()
        plt.grid(True)
        plt.yscale('log')  # Use log scale for better visualization

        plt.subplot(1, 2, 2)
        plt.plot(history.history['mae'], label='Training MAE')
        plt.plot(history.history['val_mae'], label='Validation MAE')
        plt.title('Model MAE')
        plt.xlabel('Epoch')
        plt.ylabel('MAE')
        plt.legend()
        plt.grid(True)

        plt.tight_layout()
        plt.show()

    def make_predictions(self, X_test, registration_numbers):
        """Make predictions and prepare submission - FIXED"""
        print(f"\n=== MAKING PREDICTIONS ===")

        # Make predictions
        predictions_scaled = self.model.predict(X_test, verbose=1)
        print(f"Raw predictions shape: {predictions_scaled.shape}")
        print(f"Raw predictions sample: {predictions_scaled[:5].flatten()}")

        # Inverse transform predictions
        predictions_log = self.scaler_target.inverse_transform(predictions_scaled.reshape(-1, 1))
        predictions = np.expm1(predictions_log).flatten()  # exp(x) - 1

        print(f"Predictions after inverse transform sample: {predictions[:5]}")

        # FIXED: Average predictions properly
        predictions_per_restaurant = len(predictions) // self.original_test_size
        print(f"Predictions per restaurant: {predictions_per_restaurant}")

        final_predictions = []
        for i in range(self.original_test_size):
            start_idx = i * predictions_per_restaurant
            end_idx = (i + 1) * predictions_per_restaurant
            restaurant_preds = predictions[start_idx:end_idx]
            avg_prediction = np.mean(restaurant_preds)
            final_predictions.append(max(0, avg_prediction))  # Ensure non-negative

        final_predictions = np.array(final_predictions)
        print(f"Final predictions shape: {final_predictions.shape}")
        print(f"Final predictions sample: {final_predictions[:5]}")
        print(f"Final predictions stats - Min: {final_predictions.min():.2f}, Max: {final_predictions.max():.2f}, Mean: {final_predictions.mean():.2f}")

        return final_predictions

    def evaluate_model(self, X_train, y_train):
        """Evaluate model performance - FIXED"""
        print(f"\n=== MODEL EVALUATION ===")

        # Make predictions on training data
        train_predictions_scaled = self.model.predict(X_train, verbose=0)
        train_predictions_log = self.scaler_target.inverse_transform(train_predictions_scaled.reshape(-1, 1))
        train_predictions = np.expm1(train_predictions_log).flatten()

        # FIXED: Get actual values by averaging since we have multiple sequences per sample
        original_train_size = len(self.processed_train)
        predictions_per_sample = len(train_predictions) // original_train_size

        # Average predictions for each original sample
        train_pred_averaged = []
        for i in range(original_train_size):
            start_idx = i * predictions_per_sample
            end_idx = (i + 1) * predictions_per_sample
            avg_pred = np.mean(train_predictions[start_idx:end_idx])
            train_pred_averaged.append(avg_pred)

        train_pred_averaged = np.array(train_pred_averaged)
        train_actual = self.processed_train['Annual Turnover'].values

        # Calculate RMSE
        train_rmse = np.sqrt(mean_squared_error(train_actual, train_pred_averaged))

        print(f"Training RMSE: {train_rmse:,.2f}")
        print(f"Target RMSE: 12,240,429.84")
        print(f"Performance: {'PASSED' if train_rmse < 12240429.84 else 'NEEDS IMPROVEMENT'}")

        # Additional diagnostics
        print(f"Actual values - Min: {train_actual.min():.2f}, Max: {train_actual.max():.2f}, Mean: {train_actual.mean():.2f}")
        print(f"Predicted values - Min: {train_pred_averaged.min():.2f}, Max: {train_pred_averaged.max():.2f}, Mean: {train_pred_averaged.mean():.2f}")

        return train_rmse

    def create_submission_file(self, predictions, registration_numbers, filename='lstm_submission.csv'):
        """Create submission file"""
        submission_df = pd.DataFrame({
            'Registration Number': registration_numbers,
            'Annual Turnover': predictions
        })

        submission_df.to_csv(filename, index=False)
        print(f"\nSubmission file saved as: {filename}")
        print(f"Submission shape: {submission_df.shape}")
        print(f"Prediction statistics:")
        print(submission_df['Annual Turnover'].describe())

        # Check for any zero predictions
        zero_predictions = (submission_df['Annual Turnover'] == 0).sum()
        print(f"Number of zero predictions: {zero_predictions}")

        return submission_df

def main():
    """Main execution function"""
    print("=== RESTAURANT ANNUAL TURNOVER PREDICTION - FIXED LSTM PIPELINE ===")

    # Initialize predictor
    predictor = RestaurantTurnoverPredictor()

    # File paths (update these to your actual file paths)
    train_path = "Train_dataset_(1)_(1)_(2)_(1).csv"
    test_path = "Test_dataset_(1)_(1)_(2)_(1).csv"

    try:
        # Load data
        train_df, test_df = predictor.load_data(train_path, test_path)

        # Explore data
        predictor.explore_data()

        # Preprocess data
        processed_train, processed_test = predictor.preprocess_data()

        # Prepare LSTM data with reduced sequence length
        X_train, y_train, X_test, n_features = predictor.prepare_lstm_data(sequence_length=5)

        # Build model
        predictor.model = predictor.build_lstm_model(
            input_shape=(X_train.shape[1], X_train.shape[2]),
            n_features=n_features
        )

        # Train model with fewer epochs initially
        history = predictor.train_model(X_train, y_train, epochs=100, batch_size=16)

        # Plot training history
        predictor.plot_training_history(history)

        # Evaluate model
        train_rmse = predictor.evaluate_model(X_train, y_train)

        # Make predictions
        test_predictions = predictor.make_predictions(X_test, processed_test['Registration Number'].values)

        # Create submission file
        submission_df = predictor.create_submission_file(
            test_predictions,
            processed_test['Registration Number'].values
        )

        print(f"\n=== PIPELINE COMPLETED SUCCESSFULLY ===")
        print(f"Final Training RMSE: {train_rmse:,.2f}")

    except FileNotFoundError:
        print("\nERROR: Data files not found!")
        print("Please ensure the following files are in your working directory:")
        print("- Train_dataset_(1)_(1)_(2)_(1).csv")
        print("- Test_dataset_(1)_(1)_(2)_(1).csv")
        print("\nYou can modify the file paths in the main() function if needed.")
    except Exception as e:
        print(f"\nERROR: {str(e)}")
        import traceback
        traceback.print_exc()

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder, RobustScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Ridge
import xgboost as xgb # Import XGBoost
# import tensorflow as tf # Removed TensorFlow
# from tensorflow.keras.models import Sequential # Removed TensorFlow
# from tensorflow.keras.layers import Dense, Dropout, BatchNormalization # Removed TensorFlow
# from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau # Removed TensorFlow
# from tensorflow.keras.optimizers import Adam # Removed TensorFlow
import warnings
warnings.filterwarnings('ignore')

# Set random seeds for reproducibility
np.random.seed(42)
# tf.random.set_seed(42) # Removed TensorFlow

class RestaurantTurnoverPredictor:
    def __init__(self):
        self.scaler_features = RobustScaler()  # Better for outliers
        self.scaler_target = RobustScaler()
        self.label_encoders = {}
        self.model = None # Removed Neural Network model
        self.rf_model = None # Store RF model
        self.ridge_model = None # Ridge is still an option, but we will focus on RF and XGBoost for ensemble
        self.xgb_model = None # Store XGBoost model
        self.use_ensemble = True # Still using ensemble

    def load_data(self, train_path, test_path):
        """Load training and test datasets"""
        self.train_df = pd.read_csv(train_path)
        self.test_df = pd.read_csv(test_path)
        print(f"Training data shape: {self.train_df.shape}")
        print(f"Test data shape: {self.test_df.shape}")
        return self.train_df, self.test_df

    def explore_data(self):
        """Perform basic data exploration"""
        print("\n=== DATA EXPLORATION ===")
        print("\nTraining Data Info:")
        print(self.train_df.info())
        print(f"\nMissing values in training data:\n{self.train_df.isnull().sum()}")
        print(f"\nTarget variable statistics:")
        print(self.train_df['Annual Turnover'].describe())

        # Check for outliers
        Q1 = self.train_df['Annual Turnover'].quantile(0.25)
        Q3 = self.train_df['Annual Turnover'].quantile(0.75)
        IQR = Q3 - Q1
        outliers = ((self.train_df['Annual Turnover'] < (Q1 - 1.5 * IQR)) |
                   (self.train_df['Annual Turnover'] > (Q3 + 1.5 * IQR))).sum()
        print(f"Number of outliers: {outliers}")

        # Visualize target distribution
        plt.figure(figsize=(15, 5))

        plt.subplot(1, 3, 1)
        plt.hist(self.train_df['Annual Turnover'], bins=50, edgecolor='black', alpha=0.7)
        plt.title('Distribution of Annual Turnover')
        plt.xlabel('Annual Turnover (INR)')
        plt.ylabel('Frequency')

        plt.subplot(1, 3, 2)
        plt.hist(np.log1p(self.train_df['Annual Turnover']), bins=50, edgecolor='black', alpha=0.7)
        plt.title('Log-transformed Annual Turnover')
        plt.xlabel('Log(Annual Turnover + 1)')
        plt.ylabel('Frequency')

        plt.subplot(1, 3, 3)
        plt.boxplot(self.train_df['Annual Turnover'])
        plt.title('Boxplot of Annual Turnover')
        plt.ylabel('Annual Turnover (INR)')

        plt.tight_layout()
        plt.show()

    def preprocess_data(self):
        """Enhanced data preprocessing"""
        print("\n=== DATA PREPROCESSING ===")

        # Combine train and test for consistent preprocessing
        combined_df = pd.concat([self.train_df, self.test_df], ignore_index=True, sort=False)

        # Handle date features
        if 'Opening Day of Restaurant' in combined_df.columns:
            combined_df['Opening Day of Restaurant'] = pd.to_datetime(
                combined_df['Opening Day of Restaurant'], errors='coerce'
            )
            # Extract useful date features
            combined_df['Opening_Year'] = combined_df['Opening Day of Restaurant'].dt.year
            combined_df['Opening_Month'] = combined_df['Opening Day of Restaurant'].dt.month
            combined_df['Opening_DayOfYear'] = combined_df['Opening Day of Restaurant'].dt.dayofyear
            combined_df['Days_Since_Opening'] = (pd.Timestamp.now() - combined_df['Opening Day of Restaurant']).dt.days

            # Create cyclical features for better representation
            combined_df['Opening_Month_Sin'] = np.sin(2 * np.pi * combined_df['Opening_Month'] / 12)
            combined_df['Opening_Month_Cos'] = np.cos(2 * np.pi * combined_df['Opening_Month'] / 12)
            combined_df['Opening_DayOfYear_Sin'] = np.sin(2 * np.pi * combined_df['Opening_DayOfYear'] / 365)
            combined_df['Opening_DayOfYear_Cos'] = np.cos(2 * np.pi * combined_df['Opening_DayOfYear'] / 365)

            combined_df = combined_df.drop('Opening Day of Restaurant', axis=1)

        # Handle missing values more intelligently
        print("Handling missing values...")

        # For numerical columns with high missing rates, create indicator variables
        high_missing_cols = ['Comedy Gigs Rating', 'Value Deals Rating', 'Live Sports Rating']
        for col in high_missing_cols:
            if col in combined_df.columns:
                combined_df[f'{col}_Missing'] = combined_df[col].isnull().astype(int)

        # Numerical columns
        numerical_cols = combined_df.select_dtypes(include=[np.number]).columns
        for col in numerical_cols:
            if col != 'Annual Turnover':
                combined_df[col] = combined_df[col].fillna(combined_df[col].median())

        # Categorical columns
        categorical_cols = combined_df.select_dtypes(include=['object']).columns
        for col in categorical_cols:
            if col != 'Registration Number':
                mode_val = combined_df[col].mode()
                if len(mode_val) > 0:
                    combined_df[col] = combined_df[col].fillna(mode_val[0])
                else:
                    combined_df[col] = combined_df[col].fillna('Unknown')

        # Enhanced feature engineering
        print("Creating enhanced features...")

        # Social media engagement score
        if 'Facebook Popularity Quotient' in combined_df.columns and 'Instagram Popularity Quotient' in combined_df.columns:
            combined_df['Social_Media_Score'] = (
                combined_df['Facebook Popularity Quotient'] +
                combined_df['Instagram Popularity Quotient']
            ) / 2
            combined_df['Social_Media_Ratio'] = combined_df['Facebook Popularity Quotient'] / (combined_df['Instagram Popularity Quotient'] + 1)

        # Service quality composite score (weighted)
        service_cols = ['Staff Responsivness', 'Hygiene Rating', 'Food Rating', 'Service']
        available_service_cols = [col for col in service_cols if col in combined_df.columns]
        if available_service_cols:
            # Weighted average (Food Rating gets higher weight)
            weights = {'Food Rating': 0.4, 'Service': 0.3, 'Hygiene Rating': 0.2, 'Staff Responsivness': 0.1}
            combined_df['Service_Quality_Score'] = 0
            total_weight = 0
            for col in available_service_cols:
                weight = weights.get(col, 0.25)
                combined_df['Service_Quality_Score'] += combined_df[col] * weight
                total_weight += weight
            combined_df['Service_Quality_Score'] /= total_weight

        # Rating consistency score
        rating_cols = ['Food Rating', 'Service', 'Hygiene Rating', 'Overall Restaurant Rating']
        available_rating_cols = [col for col in rating_cols if col in combined_df.columns]
        if len(available_rating_cols) > 1:
            combined_df['Rating_Std'] = combined_df[available_rating_cols].std(axis=1)
            combined_df['Rating_Range'] = combined_df[available_rating_cols].max(axis=1) - combined_df[available_rating_cols].min(axis=1)

        # Entertainment and ambience score
        entertainment_cols = ['Live Music Rating', 'Comedy Gigs Rating', 'Ambience', 'Lively']
        available_entertainment_cols = [col for col in entertainment_cols if col in combined_df.columns]
        if available_entertainment_cols:
            combined_df['Entertainment_Score'] = combined_df[available_entertainment_cols].mean(axis=1)

        # Infrastructure and convenience score
        infra_cols = ['Fire Audit', 'Liquor License Obtained', 'Situated in a Multi Complex',
                      'Dedicated Parking', 'Open Sitting Available']
        available_infra_cols = [col for col in infra_cols if col in combined_df.columns]
        if available_infra_cols:
            combined_df['Infrastructure_Score'] = combined_df[available_infra_cols].sum(axis=1)

        # City and location features
        if 'City' in combined_df.columns:
            city_counts = combined_df['City'].value_counts()
            combined_df['City_Frequency'] = combined_df['City'].map(city_counts)

        # Restaurant age feature (more sophisticated)
        if 'Days_Since_Opening' in combined_df.columns:
            combined_df['Restaurant_Age_Years'] = combined_df['Days_Since_Opening'] / 365.25
            combined_df['Is_New_Restaurant'] = (combined_df['Restaurant_Age_Years'] < 1).astype(int)
            combined_df['Is_Established'] = (combined_df['Restaurant_Age_Years'] > 5).astype(int)

        # Value perception ratio
        if 'Value for Money' in combined_df.columns and 'Food Rating' in combined_df.columns:
            combined_df['Value_Quality_Ratio'] = combined_df['Value for Money'] / (combined_df['Food Rating'] + 1)

        # Encode categorical variables with frequency encoding for high cardinality
        print("Encoding categorical variables...")
        high_cardinality_cols = []
        for col in categorical_cols:
            if col not in ['Registration Number'] and col in combined_df.columns:
                unique_count = combined_df[col].nunique()
                if unique_count > 20:  # High cardinality
                    # Frequency encoding
                    freq_map = combined_df[col].value_counts().to_dict()
                    combined_df[f'{col}_Frequency'] = combined_df[col].map(freq_map)
                    high_cardinality_cols.append(col)
                else:
                    # Label encoding for low cardinality
                    le = LabelEncoder()
                    combined_df[col] = le.fit_transform(combined_df[col].astype(str))
                    self.label_encoders[col] = le

        # Drop original high cardinality columns
        combined_df = combined_df.drop(columns=high_cardinality_cols)

        # Split back into train and test
        train_size = len(self.train_df)
        self.processed_train = combined_df[:train_size].copy()
        self.processed_test = combined_df[train_size:].copy()

        print(f"Processed training data shape: {self.processed_train.shape}")
        print(f"Processed test data shape: {self.processed_test.shape}")

        return self.processed_train, self.processed_test

    def prepare_data(self):
        """Prepare data for models"""
        print(f"\n=== PREPARING DATA FOR MODELS ===")

        # Prepare features
        feature_cols = [col for col in self.processed_train.columns
                       if col not in ['Registration Number', 'Annual Turnover']]

        X_train = self.processed_train[feature_cols].values
        y_train = self.processed_train['Annual Turnover'].values
        X_test = self.processed_test[feature_cols].values

        print(f"Feature matrix shape: {X_train.shape}")
        print(f"Target shape: {y_train.shape}")

        # Apply log transformation to target (for scaling and potential use in models)
        y_train_log = np.log1p(y_train)

        # Scale features and target
        X_train_scaled = self.scaler_features.fit_transform(X_train)
        X_test_scaled = self.scaler_features.transform(X_test)
        y_train_scaled = self.scaler_target.fit_transform(y_train_log.reshape(-1, 1)).flatten()

        print(f"Target statistics after log+scaling - Min: {y_train_scaled.min():.3f}, Max: {y_train_scaled.max():.3f}")

        return X_train_scaled, y_train_scaled, X_test_scaled, len(feature_cols), X_train, y_train # Return original X_train, y_train for tree models


    def create_ensemble_model(self, X_train_scaled, y_train_scaled, X_train_original, y_train_original):
        """Create ensemble with Random Forest and XGBoost"""
        print("\n=== CREATING ENSEMBLE MODEL (RF + XGBoost) ===")

        # Random Forest
        self.rf_model = RandomForestRegressor(
            n_estimators=200,
            max_depth=15,
            min_samples_split=10,
            min_samples_leaf=5,
            random_state=42,
            n_jobs=-1
        )
        self.rf_model.fit(X_train_original, y_train_original)

        # XGBoost Regressor
        self.xgb_model = xgb.XGBRegressor(
            objective='reg:squarederror',
            n_estimators=300,
            learning_rate=0.05,
            max_depth=6,
            subsample=0.8,
            colsample_bytree=0.8,
            random_state=42,
            n_jobs=-1
        )
        self.xgb_model.fit(X_train_original, y_train_original)

        # Ridge model is no longer part of this ensemble
        self.ridge_model = None

        return self.rf_model, self.xgb_model # Return only RF and XGBoost

    def make_ensemble_predictions(self, X_test_scaled, X_test_original):
        """Make predictions using ensemble (RF + XGBoost)"""
        # Random Forest predictions (uses original scale data)
        rf_pred = self.rf_model.predict(X_test_original)

         # XGBoost predictions (uses original scale data)
        xgb_pred = self.xgb_model.predict(X_test_original)

        # Simple averaging ensemble
        ensemble_pred = (rf_pred + xgb_pred) / 2.0


        return ensemble_pred, rf_pred, xgb_pred # Return ensemble and individual predictions

    def evaluate_model(self, X_train_scaled, y_train_scaled, X_train_original):
        """Evaluate model performance"""
        print(f"\n=== MODEL EVALUATION ===")

        train_actual = np.expm1(self.scaler_target.inverse_transform(y_train_scaled.reshape(-1, 1)).flatten())

        if self.use_ensemble and self.rf_model and self.xgb_model:
            # Ensemble evaluation
            train_pred_ensemble, train_pred_rf, train_pred_xgb = self.make_ensemble_predictions(
                X_train_scaled, X_train_original
            )

            # Calculate RMSE for each model
            rmse_ensemble = np.sqrt(mean_squared_error(train_actual, train_pred_ensemble))
            rmse_rf = np.sqrt(mean_squared_error(train_actual, train_pred_rf))
            rmse_xgb = np.sqrt(mean_squared_error(train_actual, train_pred_xgb))


            print(f"Ensemble (RF + XGBoost) RMSE: {rmse_ensemble:,.2f}")
            print(f"Random Forest RMSE: {rmse_rf:,.2f}")
            print(f"XGBoost RMSE: {rmse_xgb:,.2f}")


            best_rmse = rmse_ensemble
        else:
             # This case should not happen if use_ensemble is True
             print("No ensemble model available.")
             best_rmse = float('inf') # Indicate an issue


        print(f"Target RMSE: 12,240,429.84")
        print(f"Performance: {'PASSED' if best_rmse < 12240429.84 else 'NEEDS IMPROVEMENT'}")

        return best_rmse

    # Removed plot_training_history as NN is removed
    # def plot_training_history(self, history):
    #     """Plot training history"""
    #     plt.figure(figsize=(15, 5))

    #     plt.subplot(1, 2, 1)
    #     plt.plot(history.history['loss'], label='Training Loss', linewidth=2)
    #     plt.plot(history.history['val_loss'], label='Validation Loss', linewidth=2)
    #     plt.title('Model Loss (Huber)', fontsize=14)
    #     plt.xlabel('Epoch')
    #     plt.ylabel('Loss')
    #     plt.legend()
    #     plt.grid(True, alpha=0.3)
    #     plt.yscale('log')

    #     plt.subplot(1, 2, 2)
    #     plt.plot(history.history['mae'], label='Training MAE', linewidth=2)
    #     plt.plot(history.history['val_mae'], label='Validation MAE', linewidth=2)
    #     plt.title('Model MAE', fontsize=14)
    #     plt.xlabel('Epoch')
    #     plt.ylabel('MAE')
    #     plt.legend()
    #     plt.grid(True, alpha=0.3)

    #     plt.tight_layout()
    #     plt.show()


    def create_submission_file(self, predictions, registration_numbers, filename='optimized_submission.csv'):
        """Create submission file"""
        submission_df = pd.DataFrame({
            'Registration Number': registration_numbers,
            'Annual Turnover': np.round(predictions).astype(int)  # Round to integers
        })

        submission_df.to_csv(filename, index=False)
        print(f"\nSubmission file saved as: {filename}")
        print(f"Submission shape: {submission_df.shape}")
        print(f"Prediction statistics:")
        print(submission_df['Annual Turnover'].describe())

        return submission_df

def main():
    """Main execution function"""
    print("=== OPTIMIZED RESTAURANT ANNUAL TURNOVER PREDICTION ===")

    # Initialize predictor
    predictor = RestaurantTurnoverPredictor()

    # File paths
    train_path = "Train_dataset_(1)_(1)_(2)_(1).csv"
    test_path = "Test_dataset_(1)_(1)_(2)_(1).csv"

    try:
        # Load data
        train_df, test_df = predictor.load_data(train_path, test_path)

        # Explore data
        predictor.explore_data()

        # Preprocess data with enhanced features
        processed_train, processed_test = predictor.preprocess_data()

        # Prepare data for models (get original and scaled data)
        X_train_scaled, y_train_scaled, X_test_scaled, n_features, X_train_original, y_train_original = predictor.prepare_data()
        X_test_original = processed_test[[col for col in processed_test.columns if col not in ['Registration Number', 'Annual Turnover']]].values


        # Create ensemble models (RF + XGBoost)
        if predictor.use_ensemble:
            predictor.rf_model, predictor.xgb_model = predictor.create_ensemble_model(
                X_train_scaled, y_train_scaled, X_train_original, y_train_original
            )

            # Evaluate ensemble
            train_rmse = predictor.evaluate_model(X_train_scaled, y_train_scaled, X_train_original)

            # Make ensemble predictions
            test_predictions, _, _ = predictor.make_ensemble_predictions(X_test_scaled, X_test_original)
        else:
             # If not using ensemble, you might train a single model here if needed
             # For now, we are focusing on the ensemble
             print("Ensemble is not enabled.")
             test_predictions = np.zeros(len(processed_test)) # Placeholder


        # Create submission file
        submission_df = predictor.create_submission_file(
            test_predictions,
            processed_test['Registration Number'].values
        )

        print(f"\n=== PIPELINE COMPLETED SUCCESSFULLY ===")
        print(f"Final Training RMSE: {train_rmse:,.2f}")

    except FileNotFoundError:
        print("\nERROR: Data files not found!")
        print("Please ensure the following files are in your working directory:")
        print("- Train_dataset_(1)_(1)_(2)_(1).csv")
        print("- Test_dataset_(1)_(1)_(2)_(1).csv")
    except Exception as e:
        print(f"\nERROR: {str(e)}")
        import traceback
        traceback.print_exc()

if __name__ == "__main__":
    main()
```

Create more sophisticated features based on domain knowledge and correlations identified during data exploration. This could include interaction terms, polynomial features, or more advanced encoding techniques.


**Reasoning**:
Implement enhanced feature engineering within the `preprocess_data` method to create interaction terms, polynomial features, and explore more advanced encoding techniques based on the identified patterns in the residual analysis, ensuring consistency across train and test sets.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder, RobustScaler, PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Ridge
import xgboost as xgb
import warnings
warnings.filterwarnings('ignore')

# Set random seeds for reproducibility
np.random.seed(42)

# Custom Target Encoding (simplified for demonstration)
class TargetEncoder:
    def __init__(self, cols):
        self.cols = cols
        self.mapping = {}
        self.global_mean = 0

    def fit(self, X, y):
        self.global_mean = y.mean()
        for col in self.cols:
            self.mapping[col] = X.groupby(col)[y].mean().to_dict()

    def transform(self, X):
        X_encoded = X.copy()
        for col in self.cols:
            X_encoded[col] = X_encoded[col].map(self.mapping[col]).fillna(self.global_mean)
        return X_encoded

class RestaurantTurnoverPredictor:
    def __init__(self):
        self.scaler_features = RobustScaler()  # Better for outliers
        self.scaler_target = RobustScaler()
        self.label_encoders = {}
        self.target_encoders = {} # Store target encoders
        self.rf_model = None
        self.ridge_model = None
        self.xgb_model = None
        self.use_ensemble = True
        self.poly = PolynomialFeatures(degree=2, include_bias=False) # Polynomial features

    def load_data(self, train_path, test_path):
        """Load training and test datasets"""
        self.train_df = pd.read_csv(train_path)
        self.test_df = pd.read_csv(test_path)
        print(f"Training data shape: {self.train_df.shape}")
        print(f"Test data shape: {self.test_df.shape}")
        return self.train_df, self.test_df

    def explore_data(self):
        """Perform basic data exploration"""
        print("\n=== DATA EXPLORATION ===")
        print("\nTraining Data Info:")
        print(self.train_df.info())
        print(f"\nMissing values in training data:\n{self.train_df.isnull().sum()}")
        print(f"\nTarget variable statistics:")
        print(self.train_df['Annual Turnover'].describe())

        # Check for outliers
        Q1 = self.train_df['Annual Turnover'].quantile(0.25)
        Q3 = self.train_df['Annual Turnover'].quantile(0.75)
        IQR = Q3 - Q1
        outliers = ((self.train_df['Annual Turnover'] < (Q1 - 1.5 * IQR)) |
                   (self.train_df['Annual Turnover'] > (Q3 + 1.5 * IQR))).sum()
        print(f"Number of outliers: {outliers}")

        # Visualize target distribution
        plt.figure(figsize=(15, 5))

        plt.subplot(1, 3, 1)
        plt.hist(self.train_df['Annual Turnover'], bins=50, edgecolor='black', alpha=0.7)
        plt.title('Distribution of Annual Turnover')
        plt.xlabel('Annual Turnover (INR)')
        plt.ylabel('Frequency')

        plt.subplot(1, 3, 2)
        plt.hist(np.log1p(self.train_df['Annual Turnover']), bins=50, edgecolor='black', alpha=0.7)
        plt.title('Log-transformed Annual Turnover')
        plt.xlabel('Log(Annual Turnover + 1)')
        plt.ylabel('Frequency')

        plt.subplot(1, 3, 3)
        plt.boxplot(self.train_df['Annual Turnover'])
        plt.title('Boxplot of Annual Turnover')
        plt.ylabel('Annual Turnover (INR)')

        plt.tight_layout()
        plt.show()

    def preprocess_data(self):
        """Enhanced data preprocessing"""
        print("\n=== DATA PREPROCESSING ===")

        # Combine train and test for consistent preprocessing
        combined_df = pd.concat([self.train_df.drop(columns=['Annual Turnover'], errors='ignore'), self.test_df], ignore_index=True, sort=False)

        # Store original target
        train_target = self.train_df['Annual Turnover'].copy()

        # Handle date features
        if 'Opening Day of Restaurant' in combined_df.columns:
            combined_df['Opening Day of Restaurant'] = pd.to_datetime(
                combined_df['Opening Day of Restaurant'], errors='coerce'
            )
            # Extract useful date features
            combined_df['Opening_Year'] = combined_df['Opening Day of Restaurant'].dt.year
            combined_df['Opening_Month'] = combined_df['Opening Day of Restaurant'].dt.month
            combined_df['Opening_DayOfYear'] = combined_df['Opening Day of Restaurant'].dt.dayofyear
            combined_df['Days_Since_Opening'] = (pd.Timestamp.now() - combined_df['Opening Day of Restaurant']).dt.days

            # Create cyclical features for better representation
            combined_df['Opening_Month_Sin'] = np.sin(2 * np.pi * combined_df['Opening_Month'] / 12)
            combined_df['Opening_Month_Cos'] = np.cos(2 * np.pi * combined_df['Opening_Month'] / 12)
            combined_df['Opening_DayOfYear_Sin'] = np.sin(2 * np.pi * combined_df['Opening_DayOfYear'] / 365)
            combined_df['Opening_DayOfYear_Cos'] = np.cos(2 * np.pi * combined_df['Opening_DayOfYear'] / 365)

            combined_df = combined_df.drop('Opening Day of Restaurant', axis=1)

        # Handle missing values more intelligently
        print("Handling missing values...")

        # For numerical columns with high missing rates, create indicator variables
        high_missing_cols = ['Comedy Gigs Rating', 'Value Deals Rating', 'Live Sports Rating']
        for col in high_missing_cols:
            if col in combined_df.columns:
                combined_df[f'{col}_Missing'] = combined_df[col].isnull().astype(int)

        # Numerical columns
        numerical_cols = combined_df.select_dtypes(include=[np.number]).columns
        for col in numerical_cols:
            if col != 'Annual Turnover': # Ensure 'Annual Turnover' is not processed here
                combined_df[col] = combined_df[col].fillna(combined_df[col].median())

        # Categorical columns
        categorical_cols = combined_df.select_dtypes(include=['object']).columns
        for col in categorical_cols:
            if col != 'Registration Number':
                mode_val = combined_df[col].mode()
                if len(mode_val) > 0:
                    combined_df[col] = combined_df[col].fillna(mode_val[0])
                else:
                    combined_df[col] = combined_df[col].fillna('Unknown')

        # Enhanced feature engineering
        print("Creating enhanced features...")

        # Social media engagement score and ratio
        if 'Facebook Popularity Quotient' in combined_df.columns and 'Instagram Popularity Quotient' in combined_df.columns:
            combined_df['Social_Media_Score'] = (
                combined_df['Facebook Popularity Quotient'] +
                combined_df['Instagram Popularity Quotient']
            ) / 2
            combined_df['Social_Media_Ratio'] = combined_df['Facebook Popularity Quotient'] / (combined_df['Instagram Popularity Quotient'].replace(0, 1) + 1) # Avoid division by zero

        # Service quality composite score (weighted)
        service_cols = ['Staff Responsivness', 'Hygiene Rating', 'Food Rating', 'Service']
        available_service_cols = [col for col in service_cols if col in combined_df.columns]
        if available_service_cols:
            # Weighted average (Food Rating gets higher weight)
            weights = {'Food Rating': 0.4, 'Service': 0.3, 'Hygiene Rating': 0.2, 'Staff Responsivness': 0.1}
            combined_df['Service_Quality_Score'] = 0
            total_weight = 0
            for col in available_service_cols:
                weight = weights.get(col, 0.25)
                combined_df['Service_Quality_Score'] += combined_df[col] * weight
                total_weight += weight
            if total_weight > 0:
                combined_df['Service_Quality_Score'] /= total_weight
            else:
                combined_df['Service_Quality_Score'] = combined_df[available_service_cols].mean(axis=1) # Fallback to simple mean


        # Rating consistency score
        rating_cols = ['Food Rating', 'Service', 'Hygiene Rating', 'Overall Restaurant Rating']
        available_rating_cols = [col for col in rating_cols if col in combined_df.columns]
        if len(available_rating_cols) > 1:
            combined_df['Rating_Std'] = combined_df[available_rating_cols].std(axis=1).fillna(0) # Fill NaN from std of single value
            combined_df['Rating_Range'] = combined_df[available_rating_cols].max(axis=1) - combined_df[available_rating_cols].min(axis=1)

        # Entertainment and ambience score
        entertainment_cols = ['Live Music Rating', 'Comedy Gigs Rating', 'Ambience', 'Lively']
        available_entertainment_cols = [col for col in entertainment_cols if col in combined_df.columns]
        if available_entertainment_cols:
            combined_df['Entertainment_Score'] = combined_df[available_entertainment_cols].mean(axis=1).fillna(combined_df[available_entertainment_cols].median(axis=1)) # Use median for missing

        # Infrastructure and convenience score
        infra_cols = ['Fire Audit', 'Liquor License Obtained', 'Situated in a Multi Complex',
                      'Dedicated Parking', 'Open Sitting Available']
        available_infra_cols = [col for col in infra_cols if col in combined_df.columns]
        if available_infra_cols:
            combined_df['Infrastructure_Score'] = combined_df[available_infra_cols].sum(axis=1)

        # City and location features
        if 'City' in combined_df.columns:
            city_counts = combined_df['City'].value_counts()
            combined_df['City_Frequency'] = combined_df['City'].map(city_counts)

        # Restaurant age feature (more sophisticated)
        if 'Days_Since_Opening' in combined_df.columns:
            combined_df['Restaurant_Age_Years'] = combined_df['Days_Since_Opening'] / 365.25
            combined_df['Is_New_Restaurant'] = (combined_df['Restaurant_Age_Years'] < 1).astype(int)
            combined_df['Is_Established'] = (combined_df['Restaurant_Age_Years'] > 5).astype(int)

        # Value perception ratio
        if 'Value for Money' in combined_df.columns and 'Food Rating' in combined_df.columns:
            combined_df['Value_Quality_Ratio'] = combined_df['Value for Money'] / (combined_df['Food Rating'].replace(0, 1) + 1) # Avoid division by zero


        # *** NEW FEATURE ENGINEERING ***

        # 1. Interaction Terms (Examples based on potential domain knowledge)
        print("Creating interaction terms...")
        interaction_features = []
        if 'Social_Media_Score' in combined_df.columns and 'Restaurant Location' in combined_df.columns:
             combined_df['SocialMedia_x_Location'] = combined_df['Social_Media_Score'] * combined_df['Restaurant Location']
             interaction_features.append('SocialMedia_x_Location')
        if 'Service_Quality_Score' in combined_df.columns and 'Restaurant Type' in combined_df.columns:
             combined_df['ServiceQuality_x_RestaurantType'] = combined_df['Service_Quality_Score'] * combined_df['Restaurant Type']
             interaction_features.append('ServiceQuality_x_RestaurantType')
        if 'Infrastructure_Score' in combined_df.columns and 'Restaurant City Tier' in combined_df.columns:
             combined_df['Infrastructure_x_CityTier'] = combined_df['Infrastructure_Score'] * combined_df['Restaurant City Tier']
             interaction_features.append('Infrastructure_x_CityTier')
        if 'Restaurant_Age_Years' in combined_df.columns and 'Resturant Tier' in combined_df.columns:
             combined_df['Age_x_Tier'] = combined_df['Restaurant_Age_Years'] * combined_df['Resturant Tier']
             interaction_features.append('Age_x_Tier')


        # 2. Polynomial Features for highly correlated numerical features
        print("Creating polynomial features...")
        # Identify numerical columns likely to have non-linear relationships (e.g., ratings, popularity)
        polynomial_candidates = [
            'Facebook Popularity Quotient', 'Instagram Popularity Quotient',
            'Restaurant Zomato Rating', 'Value for Money', 'Food Rating',
            'Overall Restaurant Rating', 'Social_Media_Score', 'Service_Quality_Score',
            'Entertainment_Score', 'Restaurant_Age_Years'
        ]
        # Filter for columns actually present in the DataFrame
        polynomial_candidates = [col for col in polynomial_candidates if col in combined_df.columns]

        if polynomial_candidates:
            combined_df_poly = combined_df[polynomial_candidates].copy()
            # Handle potential NaNs before fitting polynomial features
            combined_df_poly = combined_df_poly.fillna(combined_df_poly.median())

            # Fit on the combined data to ensure consistent transformation
            self.poly.fit(combined_df_poly)
            poly_features = self.poly.transform(combined_df_poly)

            # Create a DataFrame for polynomial features with meaningful names
            poly_feature_names = self.poly.get_feature_names_out(polynomial_candidates)
            poly_df = pd.DataFrame(poly_features, columns=poly_feature_names, index=combined_df.index)

            # Drop original polynomial candidate columns from combined_df to avoid duplication
            combined_df = combined_df.drop(columns=polynomial_candidates)

            # Concatenate the new polynomial features
            combined_df = pd.concat([combined_df, poly_df], axis=1)
            print(f"Added {poly_df.shape[1]} polynomial features.")


        # 3. More Advanced Encoding (Target Encoding for high cardinality, One-Hot for others)
        print("Applying advanced encoding...")

        # Identify categorical columns again after potential date/feature engineering
        categorical_cols_updated = combined_df.select_dtypes(include=['object']).columns

        # Target Encode high cardinality columns on the training data first to avoid leakage
        high_cardinality_cols = []
        low_cardinality_cols = []
        for col in categorical_cols_updated:
            if col not in ['Registration Number'] and col in combined_df.columns:
                unique_count = combined_df[col].nunique()
                if unique_count > 50 and self.train_df.shape[0] > 0: # Threshold for target encoding, ensure training data exists
                    high_cardinality_cols.append(col)
                elif col != 'Registration Number': # Use One-Hot for low cardinality
                    low_cardinality_cols.append(col)


        # Apply Target Encoding (fit on training data, transform on combined)
        if high_cardinality_cols and self.train_df.shape[0] > 0:
            print(f"Applying Target Encoding to: {high_cardinality_cols}")
            # Split combined_df for target encoding fitting
            train_combined_part = combined_df.iloc[:len(self.train_df)].copy()

            for col in high_cardinality_cols:
                encoder = TargetEncoder(cols=[col])
                # Fit encoder on training data (using original target)
                encoder.fit(train_combined_part, train_target)
                self.target_encoders[col] = encoder # Store the fitted encoder

                # Transform the entire combined_df
                combined_df[f'{col}_TargetEncoded'] = encoder.transform(combined_df[[col]])

            # Drop original high cardinality columns
            combined_df = combined_df.drop(columns=high_cardinality_cols)


        # Apply One-Hot Encoding for low cardinality columns
        if low_cardinality_cols:
            print(f"Applying One-Hot Encoding to: {low_cardinality_cols}")
            # Convert to string to avoid issues with mixed types before OHE
            for col in low_cardinality_cols:
                combined_df[col] = combined_df[col].astype(str)

            combined_df = pd.get_dummies(combined_df, columns=low_cardinality_cols, drop_first=True) # Drop first to avoid multicollinearity


        # Ensure all columns are numeric after encoding, fill any remaining NaNs
        for col in combined_df.columns:
            if combined_df[col].dtype == 'object':
                # Try to convert to numeric if possible, otherwise drop or handle specifically
                try:
                    combined_df[col] = pd.to_numeric(combined_df[col])
                except ValueError:
                    # Handle non-numeric object columns that weren't encoded (shouldn't happen if logic is correct)
                    print(f"Warning: Column {col} remains object type after encoding.")
                    # Drop or decide how to handle
                    # combined_df = combined_df.drop(columns=[col])


        combined_df = combined_df.fillna(combined_df.median()) # Final fillna after all engineering


        # Split back into train and test
        train_size = len(self.train_df)
        self.processed_train = combined_df[:train_size].copy()
        self.processed_test = combined_df[train_size:].copy()

        # Add the original target back to the processed train df for convenience in model preparation
        self.processed_train['Annual Turnover'] = train_target


        print(f"Processed training data shape: {self.processed_train.shape}")
        print(f"Processed test data shape: {self.processed_test.shape}")

        return self.processed_train, self.processed_test

    def prepare_data(self):
        """Prepare data for models"""
        print(f"\n=== PREPARING DATA FOR MODELS ===")

        # Prepare features (exclude Registration Number and Annual Turnover from features)
        feature_cols = [col for col in self.processed_train.columns
                       if col not in ['Registration Number', 'Annual Turnover']]

        X_train = self.processed_train[feature_cols].values
        y_train = self.processed_train['Annual Turnover'].values
        X_test = self.processed_test[feature_cols].values

        print(f"Feature matrix shape: {X_train.shape}")
        print(f"Target shape: {y_train.shape}")

        # Apply log transformation to target (for scaling and potential use in models)
        y_train_log = np.log1p(y_train)

        # Scale features and target
        X_train_scaled = self.scaler_features.fit_transform(X_train)
        X_test_scaled = self.scaler_features.transform(X_test)
        y_train_scaled = self.scaler_target.fit_transform(y_train_log.reshape(-1, 1)).flatten()

        print(f"Target statistics after log+scaling - Min: {y_train_scaled.min():.3f}, Max: {y_train_scaled.max():.3f}")

        # Return original X_train, y_train for tree models that don't need scaling
        return X_train_scaled, y_train_scaled, X_test_scaled, len(feature_cols), X_train, y_train


    def create_ensemble_model(self, X_train_scaled, y_train_scaled, X_train_original, y_train_original):
        """Create ensemble with Random Forest and XGBoost"""
        print("\n=== CREATING ENSEMBLE MODEL (RF + XGBoost) ===")

        # Random Forest (uses original scale data)
        self.rf_model = RandomForestRegressor(
            n_estimators=300, # Increased estimators
            max_depth=20,     # Increased depth
            min_samples_split=5, # Reduced samples split
            min_samples_leaf=3, # Reduced samples leaf
            random_state=42,
            n_jobs=-1
        )
        self.rf_model.fit(X_train_original, y_train_original)

        # XGBoost Regressor (uses original scale data)
        self.xgb_model = xgb.XGBRegressor(
            objective='reg:squarederror',
            n_estimators=500, # Increased estimators
            learning_rate=0.03, # Reduced learning rate
            max_depth=7,        # Increased depth
            subsample=0.85,     # Slightly higher subsample
            colsample_bytree=0.85, # Slightly higher colsample
            random_state=42,
            n_jobs=-1
        )
        self.xgb_model.fit(X_train_original, y_train_original)

        # Ridge model is not part of this ensemble
        self.ridge_model = None

        return self.rf_model, self.xgb_model

    def make_ensemble_predictions(self, X_test_scaled, X_test_original):
        """Make predictions using ensemble (RF + XGBoost)"""
        # Random Forest predictions (uses original scale data)
        rf_pred = self.rf_model.predict(X_test_original)

         # XGBoost predictions (uses original scale data)
        xgb_pred = self.xgb_model.predict(X_test_original)

        # Simple averaging ensemble
        ensemble_pred = (rf_pred + xgb_pred) / 2.0


        return ensemble_pred, rf_pred, xgb_pred # Return ensemble and individual predictions

    def evaluate_model(self, X_train_scaled, y_train_scaled, X_train_original):
        """Evaluate model performance"""
        print(f"\n=== MODEL EVALUATION ===")

        # Inverse transform scaled target back to original scale
        train_actual_log = self.scaler_target.inverse_transform(y_train_scaled.reshape(-1, 1)).flatten()
        train_actual = np.expm1(train_actual_log)


        if self.use_ensemble and self.rf_model and self.xgb_model:
            # Ensemble evaluation
            train_pred_ensemble, train_pred_rf, train_pred_xgb = self.make_ensemble_predictions(
                X_train_scaled, X_train_original
            )

            # Calculate RMSE for each model
            rmse_ensemble = np.sqrt(mean_squared_error(train_actual, train_pred_ensemble))
            rmse_rf = np.sqrt(mean_squared_error(train_actual, train_pred_rf))
            rmse_xgb = np.sqrt(mean_squared_error(train_actual, train_pred_xgb))


            print(f"Ensemble (RF + XGBoost) RMSE: {rmse_ensemble:,.2f}")
            print(f"Random Forest RMSE: {rmse_rf:,.2f}")
            print(f"XGBoost RMSE: {rmse_xgb:,.2f}")


            best_rmse = rmse_ensemble
        else:
             # This case should not happen if use_ensemble is True
             print("No ensemble model available.")
             best_rmse = float('inf') # Indicate an issue


        print(f"Target RMSE: 12,240,429.84")
        print(f"Performance: {'PASSED' if best_rmse < 12240429.84 else 'NEEDS IMPROVEMENT'}")

        return best_rmse

    def create_submission_file(self, predictions, registration_numbers, filename='optimized_submission.csv'):
        """Create submission file"""
        # Ensure predictions are non-negative and round
        predictions[predictions < 0] = 0
        submission_df = pd.DataFrame({
            'Registration Number': registration_numbers,
            'Annual Turnover': np.round(predictions).astype(int)
        })

        submission_df.to_csv(filename, index=False)
        print(f"\nSubmission file saved as: {filename}")
        print(f"Submission shape: {submission_df.shape}")
        print(f"Prediction statistics:")
        print(submission_df['Annual Turnover'].describe())

        return submission_df

def main():
    """Main execution function"""
    print("=== OPTIMIZED RESTAURANT ANNUAL TURNOVER PREDICTION ===")

    # Initialize predictor
    predictor = RestaurantTurnoverPredictor()

    # File paths
    train_path = "Train_dataset_(1)_(1)_(2)_(1).csv"
    test_path = "Test_dataset_(1)_(1)_(2)_(1).csv"

    try:
        # Load data
        train_df, test_df = predictor.load_data(train_path, test_path)

        # Explore data
        predictor.explore_data()

        # Preprocess data with enhanced features
        processed_train, processed_test = predictor.preprocess_data()

        # Prepare data for models (get original and scaled data)
        X_train_scaled, y_train_scaled, X_test_scaled, n_features, X_train_original, y_train_original = predictor.prepare_data()
        # Ensure X_test_original is created correctly from processed_test
        X_test_original = processed_test[[col for col in processed_test.columns if col not in ['Registration Number', 'Annual Turnover']]].values


        # Create ensemble models (RF + XGBoost)
        if predictor.use_ensemble:
            predictor.rf_model, predictor.xgb_model = predictor.create_ensemble_model(
                X_train_scaled, y_train_scaled, X_train_original, y_train_original
            )

            # Evaluate ensemble on the training set
            train_rmse = predictor.evaluate_model(X_train_scaled, y_train_scaled, X_train_original)

            # Make ensemble predictions on the test set
            test_predictions, _, _ = predictor.make_ensemble_predictions(X_test_scaled, X_test_original)
        else:
             print("Ensemble is not enabled.")
             test_predictions = np.zeros(len(processed_test)) # Placeholder


        # Create submission file
        submission_df = predictor.create_submission_file(
            test_predictions,
            processed_test['Registration Number'].values
        )

        print(f"\n=== PIPELINE COMPLETED SUCCESSFULLY ===")
        print(f"Final Training RMSE: {train_rmse:,.2f}")

        return predictor # Return the predictor object

    except FileNotFoundError:
        print("\nERROR: Data files not found!")
        print("Please ensure the following files are in your working directory:")
        print("- Train_dataset_(1)_(1)_(2)_(1).csv")
        print("- Test_dataset_(1)_(1)_(2)_(1).csv")
        return None # Return None on error
    except Exception as e:
        print(f"\nERROR: {str(e)}")
        import traceback
        traceback.print_exc()
        return None # Return None on error

# Execute the main function to run the pipeline with enhanced features
predictor = main()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder, RobustScaler, PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Ridge
import xgboost as xgb
import warnings
warnings.filterwarnings('ignore')

# Set random seeds for reproducibility
np.random.seed(42)

# Custom Target Encoding (simplified for demonstration)
class TargetEncoder:
    def __init__(self, cols):
        self.cols = cols
        self.mapping = {}
        self.global_mean = 0

    def fit(self, X, y):
        self.global_mean = y.mean()
        for col in self.cols:
            self.mapping[col] = X.groupby(col)[y].mean().to_dict()

    def transform(self, X):
        X_encoded = X.copy()
        for col in self.cols:
            X_encoded[col] = X_encoded[col].map(self.mapping[col]).fillna(self.global_mean)
        return X_encoded

class RestaurantTurnoverPredictor:
    def __init__(self):
        self.scaler_features = RobustScaler()  # Better for outliers
        self.scaler_target = RobustScaler()
        self.label_encoders = {}
        self.target_encoders = {} # Store target encoders
        self.rf_model = None
        self.ridge_model = None
        self.xgb_model = None
        self.use_ensemble = True
        self.poly = PolynomialFeatures(degree=2, include_bias=False) # Polynomial features

    def load_data(self, train_path, test_path):
        """Load training and test datasets"""
        self.train_df = pd.read_csv(train_path)
        self.test_df = pd.read_csv(test_path)
        print(f"Training data shape: {self.train_df.shape}")
        print(f"Test data shape: {self.test_df.shape}")
        return self.train_df, self.test_df

    def explore_data(self):
        """Perform basic data exploration"""
        print("\n=== DATA EXPLORATION ===")
        print("\nTraining Data Info:")
        print(self.train_df.info())
        print(f"\nMissing values in training data:\n{self.train_df.isnull().sum()}")
        print(f"\nTarget variable statistics:")
        print(self.train_df['Annual Turnover'].describe())

        # Check for outliers
        Q1 = self.train_df['Annual Turnover'].quantile(0.25)
        Q3 = self.train_df['Annual Turnover'].quantile(0.75)
        IQR = Q3 - Q1
        outliers = ((self.train_df['Annual Turnover'] < (Q1 - 1.5 * IQR)) |
                   (self.train_df['Annual Turnover'] > (Q3 + 1.5 * IQR))).sum()
        print(f"Number of outliers: {outliers}")

        # Visualize target distribution
        plt.figure(figsize=(15, 5))

        plt.subplot(1, 3, 1)
        plt.hist(self.train_df['Annual Turnover'], bins=50, edgecolor='black', alpha=0.7)
        plt.title('Distribution of Annual Turnover')
        plt.xlabel('Annual Turnover (INR)')
        plt.ylabel('Frequency')

        plt.subplot(1, 3, 2)
        plt.hist(np.log1p(self.train_df['Annual Turnover']), bins=50, edgecolor='black', alpha=0.7)
        plt.title('Log-transformed Annual Turnover')
        plt.xlabel('Log(Annual Turnover + 1)')
        plt.ylabel('Frequency')

        plt.subplot(1, 3, 3)
        plt.boxplot(self.train_df['Annual Turnover'])
        plt.title('Boxplot of Annual Turnover')
        plt.ylabel('Annual Turnover (INR)')

        plt.tight_layout()
        plt.show()

    def preprocess_data(self):
        """Enhanced data preprocessing"""
        print("\n=== DATA PREPROCESSING ===")

        # Combine train and test for consistent preprocessing
        combined_df = pd.concat([self.train_df.drop(columns=['Annual Turnover'], errors='ignore'), self.test_df], ignore_index=True, sort=False)

        # Store original target
        train_target = self.train_df['Annual Turnover'].copy()

        # Handle date features
        if 'Opening Day of Restaurant' in combined_df.columns:
            combined_df['Opening Day of Restaurant'] = pd.to_datetime(
                combined_df['Opening Day of Restaurant'], errors='coerce'
            )
            # Extract useful date features
            combined_df['Opening_Year'] = combined_df['Opening Day of Restaurant'].dt.year
            combined_df['Opening_Month'] = combined_df['Opening Day of Restaurant'].dt.month
            combined_df['Opening_DayOfYear'] = combined_df['Opening Day of Restaurant'].dt.dayofyear
            combined_df['Days_Since_Opening'] = (pd.Timestamp.now() - combined_df['Opening Day of Restaurant']).dt.days

            # Create cyclical features for better representation
            combined_df['Opening_Month_Sin'] = np.sin(2 * np.pi * combined_df['Opening_Month'] / 12)
            combined_df['Opening_Month_Cos'] = np.cos(2 * np.pi * combined_df['Opening_Month'] / 12)
            combined_df['Opening_DayOfYear_Sin'] = np.sin(2 * np.pi * combined_df['Opening_DayOfYear'] / 365)
            combined_df['Opening_DayOfYear_Cos'] = np.cos(2 * np.pi * combined_df['Opening_DayOfYear'] / 365)

            combined_df = combined_df.drop('Opening Day of Restaurant', axis=1)

        # Handle missing values more intelligently
        print("Handling missing values...")

        # For numerical columns with high missing rates, create indicator variables
        high_missing_cols = ['Comedy Gigs Rating', 'Value Deals Rating', 'Live Sports Rating']
        for col in high_missing_cols:
            if col in combined_df.columns:
                combined_df[f'{col}_Missing'] = combined_df[col].isnull().astype(int)

        # Numerical columns
        numerical_cols = combined_df.select_dtypes(include=[np.number]).columns
        for col in numerical_cols:
            if col != 'Annual Turnover': # Ensure 'Annual Turnover' is not processed here
                combined_df[col] = combined_df[col].fillna(combined_df[col].median())

        # Categorical columns - Initial handling before specific encoding
        categorical_cols = combined_df.select_dtypes(include=['object']).columns
        for col in categorical_cols:
            if col != 'Registration Number':
                mode_val = combined_df[col].mode()
                if len(mode_val) > 0:
                    combined_df[col] = combined_df[col].fillna(mode_val[0])
                else:
                    combined_df[col] = combined_df[col].fillna('Unknown')

        # --- Apply preliminary encoding for interaction terms if needed ---
        # Ensure columns used in multiplications are numeric
        cols_for_interaction_check = ['Restaurant Location', 'Restaurant Type', 'Restaurant City Tier', 'Resturant Tier']
        for col in cols_for_interaction_check:
            if col in combined_df.columns and combined_df[col].dtype == 'object':
                 # Apply Label Encoding temporarily for interaction if not high cardinality
                 if combined_df[col].nunique() <= 50: # Use a reasonable threshold
                     le = LabelEncoder()
                     combined_df[col] = le.fit_transform(combined_df[col].astype(str))
                     # Store this temp encoder if needed later, or re-encode
                     # For now, assume subsequent encoding will handle it properly
                 else:
                      # Handle high cardinality object columns before interaction
                      pass # These will be handled by Target Encoding later

        # Re-select numerical and categorical columns after potential temp encoding
        numerical_cols = combined_df.select_dtypes(include=[np.number]).columns
        categorical_cols_updated = combined_df.select_dtypes(include=['object']).columns


        # Enhanced feature engineering
        print("Creating enhanced features...")

        # Social media engagement score and ratio
        if 'Facebook Popularity Quotient' in combined_df.columns and 'Instagram Popularity Quotient' in combined_df.columns:
            combined_df['Social_Media_Score'] = (
                combined_df['Facebook Popularity Quotient'] +
                combined_df['Instagram Popularity Quotient']
            ) / 2
            # Ensure denominator is not zero; fillna after calculation
            combined_df['Social_Media_Ratio'] = combined_df['Facebook Popularity Quotient'] / (combined_df['Instagram Popularity Quotient'].replace(0, 1) + 1)
            combined_df['Social_Media_Ratio'] = combined_df['Social_Media_Ratio'].fillna(combined_df['Social_Media_Ratio'].median())


        # Service quality composite score (weighted)
        service_cols = ['Staff Responsivness', 'Hygiene Rating', 'Food Rating', 'Service']
        available_service_cols = [col for col in service_cols if col in combined_df.columns]
        if available_service_cols:
            # Weighted average (Food Rating gets higher weight)
            weights = {'Food Rating': 0.4, 'Service': 0.3, 'Hygiene Rating': 0.2, 'Staff Responsivness': 0.1}
            combined_df['Service_Quality_Score'] = 0
            total_weight = 0
            for col in available_service_cols:
                if col in combined_df.columns: # Double check column exists
                    weight = weights.get(col, 0.25)
                    combined_df['Service_Quality_Score'] += combined_df[col] * weight
                    total_weight += weight
            if total_weight > 0:
                combined_df['Service_Quality_Score'] /= total_weight
            else:
                # Handle case where none of the service columns exist
                combined_df['Service_Quality_Score'] = 0 # Or use a default value
            combined_df['Service_Quality_Score'] = combined_df['Service_Quality_Score'].fillna(combined_df['Service_Quality_Score'].median())


        # Rating consistency score
        rating_cols = ['Food Rating', 'Service', 'Hygiene Rating', 'Overall Restaurant Rating']
        available_rating_cols = [col for col in rating_cols if col in combined_df.columns]
        if len(available_rating_cols) > 1:
            combined_df['Rating_Std'] = combined_df[available_rating_cols].std(axis=1).fillna(0) # Fill NaN from std of single value
            combined_df['Rating_Range'] = combined_df[available_rating_cols].max(axis=1) - combined_df[available_rating_cols].min(axis=1)
            combined_df['Rating_Std'] = combined_df['Rating_Std'].fillna(combined_df['Rating_Std'].median())
            combined_df['Rating_Range'] = combined_df['Rating_Range'].fillna(combined_df['Rating_Range'].median())


        # Entertainment and ambience score
        entertainment_cols = ['Live Music Rating', 'Comedy Gigs Rating', 'Ambience', 'Lively']
        available_entertainment_cols = [col for col in entertainment_cols if col in combined_df.columns]
        if available_entertainment_cols:
            combined_df['Entertainment_Score'] = combined_df[available_entertainment_cols].mean(axis=1).fillna(combined_df[available_entertainment_cols].median(axis=1)) # Use median for missing
            combined_df['Entertainment_Score'] = combined_df['Entertainment_Score'].fillna(combined_df['Entertainment_Score'].median())


        # Infrastructure and convenience score
        infra_cols = ['Fire Audit', 'Liquor License Obtained', 'Situated in a Multi Complex',
                      'Dedicated Parking', 'Open Sitting Available']
        available_infra_cols = [col for col in infra_cols if col in combined_df.columns]
        if available_infra_cols:
            combined_df['Infrastructure_Score'] = combined_df[available_infra_cols].sum(axis=1)
            combined_df['Infrastructure_Score'] = combined_df['Infrastructure_Score'].fillna(combined_df['Infrastructure_Score'].median())


        # City and location features
        if 'City' in combined_df.columns:
            city_counts = combined_df['City'].value_counts()
            combined_df['City_Frequency'] = combined_df['City'].map(city_counts).fillna(0) # Fill NaN for cities only in test


        # Restaurant age feature (more sophisticated)
        if 'Days_Since_Opening' in combined_df.columns:
            combined_df['Restaurant_Age_Years'] = combined_df['Days_Since_Opening'] / 365.25
            combined_df['Is_New_Restaurant'] = (combined_df['Restaurant_Age_Years'] < 1).astype(int)
            combined_df['Is_Established'] = (combined_df['Restaurant_Age_Years'] > 5).astype(int)
            combined_df['Restaurant_Age_Years'] = combined_df['Restaurant_Age_Years'].fillna(combined_df['Restaurant_Age_Years'].median())


        # Value perception ratio
        if 'Value for Money' in combined_df.columns and 'Food Rating' in combined_df.columns:
            # Ensure denominator is not zero; fillna after calculation
            combined_df['Value_Quality_Ratio'] = combined_df['Value for Money'] / (combined_df['Food Rating'].replace(0, 1) + 1)
            combined_df['Value_Quality_Ratio'] = combined_df['Value_Quality_Ratio'].fillna(combined_df['Value_Quality_Ratio'].median())


        # *** NEW FEATURE ENGINEERING ***

        # 1. Interaction Terms (Examples based on potential domain knowledge)
        # Ensure columns are numeric before interaction
        if 'Social_Media_Score' in combined_df.columns and 'Restaurant Location' in combined_df.columns and pd.api.types.is_numeric_dtype(combined_df['Restaurant Location']):
             print("Creating SocialMedia_x_Location interaction...")
             combined_df['SocialMedia_x_Location'] = combined_df['Social_Media_Score'] * combined_df['Restaurant Location']
        if 'Service_Quality_Score' in combined_df.columns and 'Restaurant Type' in combined_df.columns and pd.api.types.is_numeric_dtype(combined_df['Restaurant Type']):
             print("Creating ServiceQuality_x_RestaurantType interaction...")
             combined_df['ServiceQuality_x_RestaurantType'] = combined_df['Service_Quality_Score'] * combined_df['Restaurant Type']
        if 'Infrastructure_Score' in combined_df.columns and 'Restaurant City Tier' in combined_df.columns and pd.api.types.is_numeric_dtype(combined_df['Restaurant City Tier']):
             print("Creating Infrastructure_x_CityTier interaction...")
             combined_df['Infrastructure_x_CityTier'] = combined_df['Infrastructure_Score'] * combined_df['Restaurant City Tier']
        if 'Restaurant_Age_Years' in combined_df.columns and 'Resturant Tier' in combined_df.columns and pd.api.types.is_numeric_dtype(combined_df['Resturant Tier']):
             print("Creating Age_x_Tier interaction...")
             combined_df['Age_x_Tier'] = combined_df['Restaurant_Age_Years'] * combined_df['Resturant Tier']


        # 2. Polynomial Features for highly correlated numerical features
        print("Creating polynomial features...")
        # Identify numerical columns likely to have non-linear relationships (e.g., ratings, popularity)
        # Re-include columns that might have been temporarily encoded if they were originally numeric
        polynomial_candidates = [
            'Facebook Popularity Quotient', 'Instagram Popularity Quotient',
            'Restaurant Zomato Rating', 'Value for Money', 'Food Rating',
            'Overall Restaurant Rating', 'Social_Media_Score', 'Service_Quality_Score',
            'Entertainment_Score', 'Restaurant_Age_Years', 'Restaurant City Tier', 'Resturant Tier' # Add others
        ]
        # Filter for columns actually present and numeric
        polynomial_candidates = [col for col in polynomial_candidates if col in combined_df.columns and pd.api.types.is_numeric_dtype(combined_df[col])]

        if polynomial_candidates:
            combined_df_poly = combined_df[polynomial_candidates].copy()
            # Handle potential NaNs before fitting polynomial features (using median from combined)
            combined_df_poly = combined_df_poly.fillna(combined_df_poly.median())

            # Fit on the combined data to ensure consistent transformation
            self.poly.fit(combined_df_poly)
            poly_features = self.poly.transform(combined_df_poly)

            # Create a DataFrame for polynomial features with meaningful names
            poly_feature_names = self.poly.get_feature_names_out(polynomial_candidates)
            poly_df = pd.DataFrame(poly_features, columns=poly_feature_names, index=combined_df.index)

            # Drop original polynomial candidate columns from combined_df to avoid duplication
            combined_df = combined_df.drop(columns=polynomial_candidates)

            # Concatenate the new polynomial features
            combined_df = pd.concat([combined_df, poly_df], axis=1)
            print(f"Added {poly_df.shape[1]} polynomial features.")


        # 3. More Advanced Encoding (Target Encoding for high cardinality, One-Hot for others)
        print("Applying advanced encoding...")

        # Identify categorical columns again after potential date/feature engineering
        categorical_cols_updated = combined_df.select_dtypes(include=['object']).columns

        # Target Encode high cardinality columns on the training data first to avoid leakage
        high_cardinality_cols = []
        low_cardinality_cols = []
        for col in categorical_cols_updated:
            if col not in ['Registration Number'] and col in combined_df.columns:
                unique_count = combined_df[col].nunique()
                # Use a threshold for target encoding vs one-hot, consider data size
                if unique_count > 50 and self.train_df.shape[0] > 0: # Threshold for target encoding, ensure training data exists
                    high_cardinality_cols.append(col)
                elif col != 'Registration Number':
                    low_cardinality_cols.append(col)


        # Apply Target Encoding (fit on training data, transform on combined)
        if high_cardinality_cols and self.train_df.shape[0] > 0:
            print(f"Applying Target Encoding to: {high_cardinality_cols}")
            # Split combined_df for target encoding fitting
            train_combined_part = combined_df.iloc[:len(self.train_df)].copy()

            for col in high_cardinality_cols:
                encoder = TargetEncoder(cols=[col])
                # Fit encoder on training data (using original target)
                encoder.fit(train_combined_part, train_target)
                self.target_encoders[col] = encoder # Store the fitted encoder

                # Transform the entire combined_df
                combined_df[f'{col}_TargetEncoded'] = encoder.transform(combined_df[[col]])

            # Drop original high cardinality columns
            combined_df = combined_df.drop(columns=high_cardinality_cols)


        # Apply One-Hot Encoding for low cardinality columns
        if low_cardinality_cols:
            print(f"Applying One-Hot Encoding to: {low_cardinality_cols}")
            # Convert to string to avoid issues with mixed types before OHE
            for col in low_cardinality_cols:
                combined_df[col] = combined_df[col].astype(str)

            combined_df = pd.get_dummies(combined_df, columns=low_cardinality_cols, drop_first=True) # Drop first to avoid multicollinearity


        # Ensure all columns are numeric after encoding, fill any remaining NaNs
        for col in combined_df.columns:
            if combined_df[col].dtype == 'object':
                # Try to convert to numeric if possible, otherwise drop or handle specifically
                try:
                    combined_df[col] = pd.to_numeric(combined_df[col])
                except ValueError:
                    # Handle non-numeric object columns that weren't encoded
                    print(f"Warning: Column {col} remains object type after encoding. Attempting to drop.")
                    combined_df = combined_df.drop(columns=[col]) # Drop if conversion fails


        combined_df = combined_df.fillna(combined_df.median()) # Final fillna after all engineering


        # Split back into train and test
        train_size = len(self.train_df)
        self.processed_train = combined_df[:train_size].copy()
        self.processed_test = combined_df[train_size:].copy()

        # Add the original target back to the processed train df for convenience in model preparation
        self.processed_train['Annual Turnover'] = train_target


        print(f"Processed training data shape: {self.processed_train.shape}")
        print(f"Processed test data shape: {self.processed_test.shape}")

        return self.processed_train, self.processed_test

    def prepare_data(self):
        """Prepare data for models"""
        print(f"\n=== PREPARING DATA FOR MODELS ===")

        # Prepare features (exclude Registration Number and Annual Turnover from features)
        feature_cols = [col for col in self.processed_train.columns
                       if col not in ['Registration Number', 'Annual Turnover']]

        X_train = self.processed_train[feature_cols].values
        y_train = self.processed_train['Annual Turnover'].values
        X_test = self.processed_test[feature_cols].values

        print(f"Feature matrix shape: {X_train.shape}")
        print(f"Target shape: {y_train.shape}")

        # Apply log transformation to target (for scaling and potential use in models)
        y_train_log = np.log1p(y_train)

        # Scale features and target
        X_train_scaled = self.scaler_features.fit_transform(X_train)
        X_test_scaled = self.scaler_features.transform(X_test)
        y_train_scaled = self.scaler_target.fit_transform(y_train_log.reshape(-1, 1)).flatten()

        print(f"Target statistics after log+scaling - Min: {y_train_scaled.min():.3f}, Max: {y_train_scaled.max():.3f}")

        # Return original X_train, y_train for tree models that don't need scaling
        return X_train_scaled, y_train_scaled, X_test_scaled, len(feature_cols), X_train, y_train


    def create_ensemble_model(self, X_train_scaled, y_train_scaled, X_train_original, y_train_original):
        """Create ensemble with Random Forest and XGBoost"""
        print("\n=== CREATING ENSEMBLE MODEL (RF + XGBoost) ===")

        # Random Forest (uses original scale data)
        self.rf_model = RandomForestRegressor(
            n_estimators=300, # Increased estimators
            max_depth=20,     # Increased depth
            min_samples_split=5, # Reduced samples split
            min_samples_leaf=3, # Reduced samples leaf
            random_state=42,
            n_jobs=-1
        )
        self.rf_model.fit(X_train_original, y_train_original)

        # XGBoost Regressor (uses original scale data)
        self.xgb_model = xgb.XGBRegressor(
            objective='reg:squarederror',
            n_estimators=500, # Increased estimators
            learning_rate=0.03, # Reduced learning rate
            max_depth=7,        # Increased depth
            subsample=0.85,     # Slightly higher subsample
            colsample_bytree=0.85, # Slightly higher colsample
            random_state=42,
            n_jobs=-1
        )
        self.xgb_model.fit(X_train_original, y_train_original)

        # Ridge model is not part of this ensemble
        self.ridge_model = None

        return self.rf_model, self.xgb_model

    def make_ensemble_predictions(self, X_test_scaled, X_test_original):
        """Make predictions using ensemble (RF + XGBoost)"""
        # Random Forest predictions (uses original scale data)
        rf_pred = self.rf_model.predict(X_test_original)

         # XGBoost predictions (uses original scale data)
        xgb_pred = self.xgb_model.predict(X_test_original)

        # Simple averaging ensemble
        ensemble_pred = (rf_pred + xgb_pred) / 2.0


        return ensemble_pred, rf_pred, xgb_pred # Return ensemble and individual predictions

    def evaluate_model(self, X_train_scaled, y_train_scaled, X_train_original):
        """Evaluate model performance"""
        print(f"\n=== MODEL EVALUATION ===")

        # Inverse transform scaled target back to original scale
        train_actual_log = self.scaler_target.inverse_transform(y_train_scaled.reshape(-1, 1)).flatten()
        train_actual = np.expm1(train_actual_log)


        if self.use_ensemble and self.rf_model and self.xgb_model:
            # Ensemble evaluation
            train_pred_ensemble, train_pred_rf, train_pred_xgb = self.make_ensemble_predictions(
                X_train_scaled, X_train_original
            )

            # Calculate RMSE for each model
            rmse_ensemble = np.sqrt(mean_squared_error(train_actual, train_pred_ensemble))
            rmse_rf = np.sqrt(mean_squared_error(train_actual, train_pred_rf))
            rmse_xgb = np.sqrt(mean_squared_error(train_actual, train_pred_xgb))


            print(f"Ensemble (RF + XGBoost) RMSE: {rmse_ensemble:,.2f}")
            print(f"Random Forest RMSE: {rmse_rf:,.2f}")
            print(f"XGBoost RMSE: {rmse_xgb:,.2f}")


            best_rmse = rmse_ensemble
        else:
             # This case should not happen if use_ensemble is True
             print("No ensemble model available.")
             best_rmse = float('inf') # Indicate an issue


        print(f"Target RMSE: 12,240,429.84")
        print(f"Performance: {'PASSED' if best_rmse < 12240429.84 else 'NEEDS IMPROVEMENT'}")

        return best_rmse

    def create_submission_file(self, predictions, registration_numbers, filename='optimized_submission.csv'):
        """Create submission file"""
        # Ensure predictions are non-negative and round
        predictions[predictions < 0] = 0
        submission_df = pd.DataFrame({
            'Registration Number': registration_numbers,
            'Annual Turnover': np.round(predictions).astype(int)
        })

        submission_df.to_csv(filename, index=False)
        print(f"\nSubmission file saved as: {filename}")
        print(f"Submission shape: {submission_df.shape}")
        print(f"Prediction statistics:")
        print(submission_df['Annual Turnover'].describe())

        return submission_df

def main():
    """Main execution function"""
    print("=== OPTIMIZED RESTAURANT ANNUAL TURNOVER PREDICTION ===")

    # Initialize predictor
    predictor = RestaurantTurnoverPredictor()

    # File paths
    train_path = "Train_dataset_(1)_(1)_(2)_(1).csv"
    test_path = "Test_dataset_(1)_(1)_(2)_(1).csv"

    try:
        # Load data
        train_df, test_df = predictor.load_data(train_path, test_path)

        # Explore data
        predictor.explore_data()

        # Preprocess data with enhanced features
        processed_train, processed_test = predictor.preprocess_data()

        # Prepare data for models (get original and scaled data)
        X_train_scaled, y_train_scaled, X_test_scaled, n_features, X_train_original, y_train_original = predictor.prepare_data()
        # Ensure X_test_original is created correctly from processed_test
        X_test_original = processed_test[[col for col in processed_test.columns if col not in ['Registration Number', 'Annual Turnover']]].values


        # Create ensemble models (RF + XGBoost)
        if predictor.use_ensemble:
            predictor.rf_model, predictor.xgb_model = predictor.create_ensemble_model(
                X_train_scaled, y_train_scaled, X_train_original, y_train_original
            )

            # Evaluate ensemble on the training set
            train_rmse = predictor.evaluate_model(X_train_scaled, y_train_scaled, X_train_original)

            # Make ensemble predictions on the test set
            test_predictions, _, _ = predictor.make_ensemble_predictions(X_test_scaled, X_test_original)
        else:
             print("Ensemble is not enabled.")
             test_predictions = np.zeros(len(processed_test)) # Placeholder


        # Create submission file
        submission_df = predictor.create_submission_file(
            test_predictions,
            processed_test['Registration Number'].values
        )

        print(f"\n=== PIPELINE COMPLETED SUCCESSFULLY ===")
        print(f"Final Training RMSE: {train_rmse:,.2f}")

        return predictor # Return the predictor object

    except FileNotFoundError:
        print("\nERROR: Data files not found!")
        print("Please ensure the following files are in your working directory:")
        print("- Train_dataset_(1)_(1)_(2)_(1).csv")
        print("- Test_dataset_(1)_(1)_(2)_(1).csv")
        return None # Return None on error
    except Exception as e:
        print(f"\nERROR: {str(e)}")
        import traceback
        traceback.print_exc()
        return None # Return None on error

# Execute the main function to run the pipeline with enhanced features
predictor = main()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder, RobustScaler, PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Ridge
import xgboost as xgb
import warnings
warnings.filterwarnings('ignore')

# Set random seeds for reproducibility
np.random.seed(42)

# Custom Target Encoding (corrected)
class TargetEncoder:
    def __init__(self, cols):
        self.cols = cols
        self.mapping = {}
        self.global_mean = 0

    def fit(self, X, y):
        self.global_mean = y.mean()
        for col in self.cols:
            # Corrected: Group by the specified column in X and calculate mean of y
            self.mapping[col] = X.groupby(col)[y.name].mean().to_dict() # Assuming y is a Series with a name

    def transform(self, X):
        X_encoded = X.copy()
        for col in self.cols:
            # Corrected: Map the values of the specified column in X
            X_encoded[col] = X_encoded[col].map(self.mapping[col]).fillna(self.global_mean)
        return X_encoded

class RestaurantTurnoverPredictor:
    def __init__(self):
        self.scaler_features = RobustScaler()  # Better for outliers
        self.scaler_target = RobustScaler()
        self.label_encoders = {}
        self.target_encoders = {} # Store target encoders
        self.rf_model = None
        self.ridge_model = None
        self.xgb_model = None
        self.use_ensemble = True
        self.poly = PolynomialFeatures(degree=2, include_bias=False) # Polynomial features

    def load_data(self, train_path, test_path):
        """Load training and test datasets"""
        self.train_df = pd.read_csv(train_path)
        self.test_df = pd.read_csv(test_path)
        print(f"Training data shape: {self.train_df.shape}")
        print(f"Test data shape: {self.test_df.shape}")
        return self.train_df, self.test_df

    def explore_data(self):
        """Perform basic data exploration"""
        print("\n=== DATA EXPLORATION ===")
        print("\nTraining Data Info:")
        print(self.train_df.info())
        print(f"\nMissing values in training data:\n{self.train_df.isnull().sum()}")
        print(f"\nTarget variable statistics:")
        print(self.train_df['Annual Turnover'].describe())

        # Check for outliers
        Q1 = self.train_df['Annual Turnover'].quantile(0.25)
        Q3 = self.train_df['Annual Turnover'].quantile(0.75)
        IQR = Q3 - Q1
        outliers = ((self.train_df['Annual Turnover'] < (Q1 - 1.5 * IQR)) |
                   (self.train_df['Annual Turnover'] > (Q3 + 1.5 * IQR))).sum()
        print(f"Number of outliers: {outliers}")

        # Visualize target distribution
        plt.figure(figsize=(15, 5))

        plt.subplot(1, 3, 1)
        plt.hist(self.train_df['Annual Turnover'], bins=50, edgecolor='black', alpha=0.7)
        plt.title('Distribution of Annual Turnover')
        plt.xlabel('Annual Turnover (INR)')
        plt.ylabel('Frequency')

        plt.subplot(1, 3, 2)
        plt.hist(np.log1p(self.train_df['Annual Turnover']), bins=50, edgecolor='black', alpha=0.7)
        plt.title('Log-transformed Annual Turnover')
        plt.xlabel('Log(Annual Turnover + 1)')
        plt.ylabel('Frequency')

        plt.subplot(1, 3, 3)
        plt.boxplot(self.train_df['Annual Turnover'])
        plt.title('Boxplot of Annual Turnover')
        plt.ylabel('Annual Turnover (INR)')

        plt.tight_layout()
        plt.show()

    def preprocess_data(self):
        """Enhanced data preprocessing"""
        print("\n=== DATA PREPROCESSING ===")

        # Combine train and test for consistent preprocessing
        combined_df = pd.concat([self.train_df.drop(columns=['Annual Turnover'], errors='ignore'), self.test_df], ignore_index=True, sort=False)

        # Store original target
        train_target = self.train_df['Annual Turnover'].copy()

        # Handle date features
        if 'Opening Day of Restaurant' in combined_df.columns:
            combined_df['Opening Day of Restaurant'] = pd.to_datetime(
                combined_df['Opening Day of Restaurant'], errors='coerce'
            )
            # Extract useful date features
            combined_df['Opening_Year'] = combined_df['Opening Day of Restaurant'].dt.year
            combined_df['Opening_Month'] = combined_df['Opening Day of Restaurant'].dt.month
            combined_df['Opening_DayOfYear'] = combined_df['Opening Day of Restaurant'].dt.dayofyear
            combined_df['Days_Since_Opening'] = (pd.Timestamp.now() - combined_df['Opening Day of Restaurant']).dt.days

            # Create cyclical features for better representation
            combined_df['Opening_Month_Sin'] = np.sin(2 * np.pi * combined_df['Opening_Month'] / 12)
            combined_df['Opening_Month_Cos'] = np.cos(2 * np.pi * combined_df['Opening_Month'] / 12)
            combined_df['Opening_DayOfYear_Sin'] = np.sin(2 * np.pi * combined_df['Opening_DayOfYear'] / 365)
            combined_df['Opening_DayOfYear_Cos'] = np.cos(2 * np.pi * combined_df['Opening_DayOfYear'] / 365)

            combined_df = combined_df.drop('Opening Day of Restaurant', axis=1)

        # Handle missing values more intelligently
        print("Handling missing values...")

        # For numerical columns with high missing rates, create indicator variables
        high_missing_cols = ['Comedy Gigs Rating', 'Value Deals Rating', 'Live Sports Rating']
        for col in high_missing_cols:
            if col in combined_df.columns:
                combined_df[f'{col}_Missing'] = combined_df[col].isnull().astype(int)

        # Numerical columns
        numerical_cols = combined_df.select_dtypes(include=[np.number]).columns
        for col in numerical_cols:
            if col != 'Annual Turnover': # Ensure 'Annual Turnover' is not processed here
                combined_df[col] = combined_df[col].fillna(combined_df[col].median())

        # Categorical columns - Initial handling before specific encoding
        categorical_cols = combined_df.select_dtypes(include=['object']).columns
        for col in categorical_cols:
            if col != 'Registration Number':
                mode_val = combined_df[col].mode()
                if len(mode_val) > 0:
                    combined_df[col] = combined_df[col].fillna(mode_val[0])
                else:
                    combined_df[col] = combined_df[col].fillna('Unknown')

        # Apply Label Encoding to specific categorical columns BEFORE interaction terms are created
        # This ensures they are numeric for multiplication
        cols_for_label_encoding = ['Restaurant Location', 'Restaurant Type', 'EndorsedBy', 'Restaurant Theme'] # Add other low-to-medium cardinality cols
        for col in cols_for_label_encoding:
            if col in combined_df.columns:
                 le = LabelEncoder()
                 combined_df[col] = le.fit_transform(combined_df[col].astype(str))
                 self.label_encoders[col] = le # Store the encoder


        # Re-select numerical and categorical columns after preliminary encoding
        numerical_cols = combined_df.select_dtypes(include=[np.number]).columns
        categorical_cols_updated = combined_df.select_dtypes(include=['object']).columns # Should mostly be 'Registration Number' and high cardinality ones now


        # Enhanced feature engineering
        print("Creating enhanced features...")

        # Social media engagement score and ratio
        if 'Facebook Popularity Quotient' in combined_df.columns and 'Instagram Popularity Quotient' in combined_df.columns:
            combined_df['Social_Media_Score'] = (
                combined_df['Facebook Popularity Quotient'] +
                combined_df['Instagram Popularity Quotient']
            ) / 2
            # Ensure denominator is not zero; fillna after calculation
            combined_df['Social_Media_Ratio'] = combined_df['Facebook Popularity Quotient'] / (combined_df['Instagram Popularity Quotient'].replace(0, 1) + 1)
            combined_df['Social_Media_Ratio'] = combined_df['Social_Media_Ratio'].fillna(combined_df['Social_Media_Ratio'].median())


        # Service quality composite score (weighted)
        service_cols = ['Staff Responsivness', 'Hygiene Rating', 'Food Rating', 'Service']
        available_service_cols = [col for col in service_cols if col in combined_df.columns]
        if available_service_cols:
            # Weighted average (Food Rating gets higher weight)
            weights = {'Food Rating': 0.4, 'Service': 0.3, 'Hygiene Rating': 0.2, 'Staff Responsivness': 0.1}
            combined_df['Service_Quality_Score'] = 0
            total_weight = 0
            for col in available_service_cols:
                if col in combined_df.columns: # Double check column exists
                    weight = weights.get(col, 0.25)
                    combined_df['Service_Quality_Score'] += combined_df[col] * weight
                    total_weight += weight
            if total_weight > 0:
                combined_df['Service_Quality_Score'] /= total_weight
            else:
                # Handle case where none of the service columns exist
                combined_df['Service_Quality_Score'] = 0 # Or use a default value
            combined_df['Service_Quality_Score'] = combined_df['Service_Quality_Score'].fillna(combined_df['Service_Quality_Score'].median())


        # Rating consistency score
        rating_cols = ['Food Rating', 'Service', 'Hygiene Rating', 'Overall Restaurant Rating']
        available_rating_cols = [col for col in rating_cols if col in combined_df.columns]
        if len(available_rating_cols) > 1:
            combined_df['Rating_Std'] = combined_df[available_rating_cols].std(axis=1).fillna(0) # Fill NaN from std of single value
            combined_df['Rating_Range'] = combined_df[available_rating_cols].max(axis=1) - combined_df[available_rating_cols].min(axis=1)
            combined_df['Rating_Std'] = combined_df['Rating_Std'].fillna(combined_df['Rating_Std'].median())
            combined_df['Rating_Range'] = combined_df['Rating_Range'].fillna(combined_df['Rating_Range'].median())


        # Entertainment and ambience score
        entertainment_cols = ['Live Music Rating', 'Comedy Gigs Rating', 'Ambience', 'Lively']
        available_entertainment_cols = [col for col in entertainment_cols if col in combined_df.columns]
        if available_entertainment_cols:
            combined_df['Entertainment_Score'] = combined_df[available_entertainment_cols].mean(axis=1).fillna(combined_df[available_entertainment_cols].median(axis=1)) # Use median for missing
            combined_df['Entertainment_Score'] = combined_df['Entertainment_Score'].fillna(combined_df['Entertainment_Score'].median())


        # Infrastructure and convenience score
        infra_cols = ['Fire Audit', 'Liquor License Obtained', 'Situated in a Multi Complex',
                      'Dedicated Parking', 'Open Sitting Available']
        available_infra_cols = [col for col in infra_cols if col in combined_df.columns]
        if available_infra_cols:
            combined_df['Infrastructure_Score'] = combined_df[available_infra_cols].sum(axis=1)
            combined_df['Infrastructure_Score'] = combined_df['Infrastructure_Score'].fillna(combined_df['Infrastructure_Score'].median())


        # City and location features
        if 'City' in combined_df.columns:
            city_counts = combined_df['City'].value_counts()
            combined_df['City_Frequency'] = combined_df['City'].map(city_counts).fillna(0) # Fill NaN for cities only in test


        # Restaurant age feature (more sophisticated)
        if 'Days_Since_Opening' in combined_df.columns:
            combined_df['Restaurant_Age_Years'] = combined_df['Days_Since_Opening'] / 365.25
            combined_df['Is_New_Restaurant'] = (combined_df['Restaurant_Age_Years'] < 1).astype(int)
            combined_df['Is_Established'] = (combined_df['Restaurant_Age_Years'] > 5).astype(int)
            combined_df['Restaurant_Age_Years'] = combined_df['Restaurant_Age_Years'].fillna(combined_df['Restaurant_Age_Years'].median())


        # Value perception ratio
        if 'Value for Money' in combined_df.columns and 'Food Rating' in combined_df.columns:
            # Ensure denominator is not zero; fillna after calculation
            combined_df['Value_Quality_Ratio'] = combined_df['Value for Money'] / (combined_df['Food Rating'].replace(0, 1) + 1)
            combined_df['Value_Quality_Ratio'] = combined_df['Value_Quality_Ratio'].fillna(combined_df['Value_Quality_Ratio'].median())


        # *** NEW FEATURE ENGINEERING ***

        # 1. Interaction Terms (Examples based on potential domain knowledge)
        # Ensure columns are numeric before interaction - handled by preliminary LE
        if 'Social_Media_Score' in combined_df.columns and 'Restaurant Location' in combined_df.columns and pd.api.types.is_numeric_dtype(combined_df['Restaurant Location']):
             print("Creating SocialMedia_x_Location interaction...")
             combined_df['SocialMedia_x_Location'] = combined_df['Social_Media_Score'] * combined_df['Restaurant Location']
        if 'Service_Quality_Score' in combined_df.columns and 'Restaurant Type' in combined_df.columns and pd.api.types.is_numeric_dtype(combined_df['Restaurant Type']):
             print("Creating ServiceQuality_x_RestaurantType interaction...")
             combined_df['ServiceQuality_x_RestaurantType'] = combined_df['Service_Quality_Score'] * combined_df['Restaurant Type']
        if 'Infrastructure_Score' in combined_df.columns and 'Restaurant City Tier' in combined_df.columns and pd.api.types.is_numeric_dtype(combined_df['Restaurant City Tier']):
             print("Creating Infrastructure_x_CityTier interaction...")
             combined_df['Infrastructure_x_CityTier'] = combined_df['Infrastructure_Score'] * combined_df['Restaurant City Tier']
        if 'Restaurant_Age_Years' in combined_df.columns and 'Resturant Tier' in combined_df.columns and pd.api.types.is_numeric_dtype(combined_df['Resturant Tier']):
             print("Creating Age_x_Tier interaction...")
             combined_df['Age_x_Tier'] = combined_df['Restaurant_Age_Years'] * combined_df['Resturant Tier']


        # 2. Polynomial Features for highly correlated numerical features
        print("Creating polynomial features...")
        # Identify numerical columns likely to have non-linear relationships (e.g., ratings, popularity)
        # Re-include columns that might have been temporarily encoded if they were originally numeric
        polynomial_candidates = [
            'Facebook Popularity Quotient', 'Instagram Popularity Quotient',
            'Restaurant Zomato Rating', 'Value for Money', 'Food Rating',
            'Overall Restaurant Rating', 'Social_Media_Score', 'Service_Quality_Score',
            'Entertainment_Score', 'Restaurant_Age_Years', 'Restaurant City Tier', 'Resturant Tier'
        ]
        # Filter for columns actually present and numeric
        polynomial_candidates = [col for col in combined_df.columns if col in polynomial_candidates and pd.api.types.is_numeric_dtype(combined_df[col])]


        if polynomial_candidates:
            combined_df_poly = combined_df[polynomial_candidates].copy()
            # Handle potential NaNs before fitting polynomial features (using median from combined)
            combined_df_poly = combined_df_poly.fillna(combined_df_poly.median())

            # Fit on the combined data to ensure consistent transformation
            self.poly.fit(combined_df_poly)
            poly_features = self.poly.transform(combined_df_poly)

            # Create a DataFrame for polynomial features with meaningful names
            poly_feature_names = self.poly.get_feature_names_out(polynomial_candidates)
            poly_df = pd.DataFrame(poly_features, columns=poly_feature_names, index=combined_df.index)

            # Drop original polynomial candidate columns from combined_df to avoid duplication
            combined_df = combined_df.drop(columns=polynomial_candidates)

            # Concatenate the new polynomial features
            combined_df = pd.concat([combined_df, poly_df], axis=1)
            print(f"Added {poly_df.shape[1]} polynomial features.")


        # 3. More Advanced Encoding (Target Encoding for high cardinality, One-Hot for others)
        print("Applying advanced encoding...")

        # Identify categorical columns again
        categorical_cols_final = combined_df.select_dtypes(include=['object']).columns

        # Target Encode high cardinality columns on the training data first to avoid leakage
        high_cardinality_cols = []
        low_cardinality_cols = []
        for col in categorical_cols_final:
            if col not in ['Registration Number'] and col in combined_df.columns:
                unique_count = combined_df[col].nunique()
                # Use a threshold for target encoding vs one-hot, consider data size
                if unique_count > 50 and self.train_df.shape[0] > 0: # Threshold for target encoding, ensure training data exists
                    high_cardinality_cols.append(col)
                elif col != 'Registration Number':
                    low_cardinality_cols.append(col)


        # Apply Target Encoding (fit on training data, transform on combined)
        if high_cardinality_cols and self.train_df.shape[0] > 0:
            print(f"Applying Target Encoding to: {high_cardinality_cols}")
            # Split combined_df for target encoding fitting
            train_combined_part = combined_df.iloc[:len(self.train_df)].copy()

            for col in high_cardinality_cols:
                encoder = TargetEncoder(cols=[col])
                # Fit encoder on training data (using original target Series)
                encoder.fit(train_combined_part, train_target)
                self.target_encoders[col] = encoder # Store the fitted encoder

                # Transform the entire combined_df
                # Need to pass the column as a DataFrame/Series for the custom encoder
                combined_df[f'{col}_TargetEncoded'] = encoder.transform(combined_df[[col]])[['Annual Turnover']] # Assuming the encoder returns a df/series named 'Annual Turnover'

            # Drop original high cardinality columns
            combined_df = combined_df.drop(columns=high_cardinality_cols)


        # Apply One-Hot Encoding for low cardinality columns
        if low_cardinality_cols:
            print(f"Applying One-Hot Encoding to: {low_cardinality_cols}")
            # Convert to string to avoid issues with mixed types before OHE
            for col in low_cardinality_cols:
                combined_df[col] = combined_df[col].astype(str)

            combined_df = pd.get_dummies(combined_df, columns=low_cardinality_cols, drop_first=True) # Drop first to avoid multicollinearity


        # Ensure all columns are numeric after encoding, fill any remaining NaNs
        for col in combined_df.columns:
            if combined_df[col].dtype == 'object':
                # Try to convert to numeric if possible, otherwise drop or handle specifically
                try:
                    combined_df[col] = pd.to_numeric(combined_df[col])
                except ValueError:
                    # Handle non-numeric object columns that weren't encoded
                    print(f"Warning: Column {col} remains object type after encoding. Attempting to drop.")
                    combined_df = combined_df.drop(columns=[col]) # Drop if conversion fails


        combined_df = combined_df.fillna(combined_df.median()) # Final fillna after all engineering


        # Split back into train and test
        train_size = len(self.train_df)
        self.processed_train = combined_df[:train_size].copy()
        self.processed_test = combined_df[train_size:].copy()

        # Add the original target back to the processed train df for convenience in model preparation
        self.processed_train['Annual Turnover'] = train_target


        print(f"Processed training data shape: {self.processed_train.shape}")
        print(f"Processed test data shape: {self.processed_test.shape}")

        return self.processed_train, self.processed_test

    def prepare_data(self):
        """Prepare data for models"""
        print(f"\n=== PREPARING DATA FOR MODELS ===")

        # Prepare features (exclude Registration Number and Annual Turnover from features)
        feature_cols = [col for col in self.processed_train.columns
                       if col not in ['Registration Number', 'Annual Turnover']]

        X_train = self.processed_train[feature_cols].values
        y_train = self.processed_train['Annual Turnover'].values
        X_test = self.processed_test[feature_cols].values

        print(f"Feature matrix shape: {X_train.shape}")
        print(f"Target shape: {y_train.shape}")

        # Apply log transformation to target (for scaling and potential use in models)
        y_train_log = np.log1p(y_train)

        # Scale features and target
        X_train_scaled = self.scaler_features.fit_transform(X_train)
        X_test_scaled = self.scaler_features.transform(X_test)
        y_train_scaled = self.scaler_target.fit_transform(y_train_log.reshape(-1, 1)).flatten()

        print(f"Target statistics after log+scaling - Min: {y_train_scaled.min():.3f}, Max: {y_train_scaled.max():.3f}")

        # Return original X_train, y_train for tree models that don't need scaling
        return X_train_scaled, y_train_scaled, X_test_scaled, len(feature_cols), X_train, y_train


    def create_ensemble_model(self, X_train_scaled, y_train_scaled, X_train_original, y_train_original):
        """Create ensemble with Random Forest and XGBoost"""
        print("\n=== CREATING ENSEMBLE MODEL (RF + XGBoost) ===")

        # Random Forest (uses original scale data)
        self.rf_model = RandomForestRegressor(
            n_estimators=300, # Increased estimators
            max_depth=20,     # Increased depth
            min_samples_split=5, # Reduced samples split
            min_samples_leaf=3, # Reduced samples leaf
            random_state=42,
            n_jobs=-1
        )
        self.rf_model.fit(X_train_original, y_train_original)

        # XGBoost Regressor (uses original scale data)
        self.xgb_model = xgb.XGBRegressor(
            objective='reg:squarederror',
            n_estimators=500, # Increased estimators
            learning_rate=0.03, # Reduced learning rate
            max_depth=7,        # Increased depth
            subsample=0.85,     # Slightly higher subsample
            colsample_bytree=0.85, # Slightly higher colsample
            random_state=42,
            n_jobs=-1
        )
        self.xgb_model.fit(X_train_original, y_train_original)

        # Ridge model is not part of this ensemble
        self.ridge_model = None

        return self.rf_model, self.xgb_model

    def make_ensemble_predictions(self, X_test_scaled, X_test_original):
        """Make predictions using ensemble (RF + XGBoost)"""
        # Random Forest predictions (uses original scale data)
        rf_pred = self.rf_model.predict(X_test_original)

         # XGBoost predictions (uses original scale data)
        xgb_pred = self.xgb_model.predict(X_test_original)

        # Simple averaging ensemble
        ensemble_pred = (rf_pred + xgb_pred) / 2.0


        return ensemble_pred, rf_pred, xgb_pred # Return ensemble and individual predictions

    def evaluate_model(self, X_train_scaled, y_train_scaled, X_train_original):
        """Evaluate model performance"""
        print(f"\n=== MODEL EVALUATION ===")

        # Inverse transform scaled target back to original scale
        train_actual_log = self.scaler_target.inverse_transform(y_train_scaled.reshape(-1, 1)).flatten()
        train_actual = np.expm1(train_actual_log)


        if self.use_ensemble and self.rf_model and self.xgb_model:
            # Ensemble evaluation
            train_pred_ensemble, train_pred_rf, train_pred_xgb = self.make_ensemble_predictions(
                X_train_scaled, X_train_original
            )

            # Calculate RMSE for each model
            rmse_ensemble = np.sqrt(mean_squared_error(train_actual, train_pred_ensemble))
            rmse_rf = np.sqrt(mean_squared_error(train_actual, train_pred_rf))
            rmse_xgb = np.sqrt(mean_squared_error(train_actual, train_pred_xgb))


            print(f"Ensemble (RF + XGBoost) RMSE: {rmse_ensemble:,.2f}")
            print(f"Random Forest RMSE: {rmse_rf:,.2f}")
            print(f"XGBoost RMSE: {rmse_xgb:,.2f}")


            best_rmse = rmse_ensemble
        else:
             # This case should not happen if use_ensemble is True
             print("No ensemble model available.")
             best_rmse = float('inf') # Indicate an issue


        print(f"Target RMSE: 12,240,429.84")
        print(f"Performance: {'PASSED' if best_rmse < 12240429.84 else 'NEEDS IMPROVEMENT'}")

        return best_rmse

    def create_submission_file(self, predictions, registration_numbers, filename='optimized_submission.csv'):
        """Create submission file"""
        # Ensure predictions are non-negative and round
        predictions[predictions < 0] = 0
        submission_df = pd.DataFrame({
            'Registration Number': registration_numbers,
            'Annual Turnover': np.round(predictions).astype(int)
        })

        submission_df.to_csv(filename, index=False)
        print(f"\nSubmission file saved as: {filename}")
        print(f"Submission shape: {submission_df.shape}")
        print(f"Prediction statistics:")
        print(submission_df['Annual Turnover'].describe())

        return submission_df

def main():
    """Main execution function"""
    print("=== OPTIMIZED RESTAURANT ANNUAL TURNOVER PREDICTION ===")

    # Initialize predictor
    predictor = RestaurantTurnoverPredictor()

    # File paths
    train_path = "Train_dataset_(1)_(1)_(2)_(1).csv"
    test_path = "Test_dataset_(1)_(1)_(2)_(1).csv"

    try:
        # Load data
        train_df, test_df = predictor.load_data(train_path, test_path)

        # Explore data
        predictor.explore_data()

        # Preprocess data with enhanced features
        processed_train, processed_test = predictor.preprocess_data()

        # Prepare data for models (get original and scaled data)
        X_train_scaled, y_train_scaled, X_test_scaled, n_features, X_train_original, y_train_original = predictor.prepare_data()
        # Ensure X_test_original is created correctly from processed_test
        X_test_original = processed_test[[col for col in processed_test.columns if col not in ['Registration Number', 'Annual Turnover']]].values


        # Create ensemble models (RF + XGBoost)
        if predictor.use_ensemble:
            predictor.rf_model, predictor.xgb_model = predictor.create_ensemble_model(
                X_train_scaled, y_train_scaled, X_train_original, y_train_original
            )

            # Evaluate ensemble on the training set
            train_rmse = predictor.evaluate_model(X_train_scaled, y_train_scaled, X_train_original)

            # Make ensemble predictions on the test set
            test_predictions, _, _ = predictor.make_ensemble_predictions(X_test_scaled, X_test_original)
        else:
             print("Ensemble is not enabled.")
             test_predictions = np.zeros(len(processed_test)) # Placeholder


        # Create submission file
        submission_df = predictor.create_submission_file(
            test_predictions,
            processed_test['Registration Number'].values
        )

        print(f"\n=== PIPELINE COMPLETED SUCCESSFULLY ===")
        print(f"Final Training RMSE: {train_rmse:,.2f}")

        return predictor # Return the predictor object

    except FileNotFoundError:
        print("\nERROR: Data files not found!")
        print("Please ensure the following files are in your working directory:")
        print("- Train_dataset_(1)_(1)_(2)_(1).csv")
        print("- Test_dataset_(1)_(1)_(2)_(1).csv")
        return None # Return None on error
    except Exception as e:
        print(f"\nERROR: {str(e)}")
        import traceback
        traceback.print_exc()
        return None # Return None on error

# Execute the main function to run the pipeline with enhanced features
predictor = main()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder, RobustScaler, PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Ridge
import xgboost as xgb
import warnings
warnings.filterwarnings('ignore')

# Set random seeds for reproducibility
np.random.seed(42)

# Custom Target Encoding (corrected)
class TargetEncoder:
    def __init__(self, cols):
        self.cols = cols
        self.mapping = {}
        self.global_mean = 0

    def fit(self, X, y):
        # Ensure y is a pandas Series
        if not isinstance(y, pd.Series):
            y = pd.Series(y)

        self.global_mean = y.mean()
        for col in self.cols:
            if col in X.columns:
                # Corrected approach: Group X by the specified column and calculate the mean of the target y
                # Create a temporary DataFrame with the categorical column from X and the target Series y
                temp_df = pd.DataFrame({col: X[col].values, 'target': y.values}) # Use .values to align indices if necessary
                self.mapping[col] = temp_df.groupby(col)['target'].mean().to_dict()

            else:
                print(f"Warning: Column '{col}' not found in DataFrame for Target Encoding fit.")
                self.mapping[col] = {} # Initialize empty mapping

    def transform(self, X):
        X_encoded = X.copy()
        for col in self.cols:
            if col in X_encoded.columns:
                # Corrected: Map the values in the specified column using the stored mapping
                # Use .get(col, {}) to handle cases where a column was in cols but not found during fit
                X_encoded[col] = X_encoded[col].map(self.mapping.get(col, {})).fillna(self.global_mean)
            else:
                print(f"Warning: Column '{col}' not found in DataFrame for Target Encoding transform.")
                # If the column is missing during transform, fill with global mean or handle differently
                # For now, we'll just skip if missing, assuming it was handled earlier if necessary.
        return X_encoded


class RestaurantTurnoverPredictor:
    def __init__(self):
        self.scaler_features = RobustScaler()  # Better for outliers
        self.scaler_target = RobustScaler()
        self.label_encoders = {}
        self.target_encoders = {} # Store target encoders
        self.rf_model = None
        self.ridge_model = None
        self.xgb_model = None
        self.use_ensemble = True
        self.poly = PolynomialFeatures(degree=2, include_bias=False) # Polynomial features

    def load_data(self, train_path, test_path):
        """Load training and test datasets"""
        self.train_df = pd.read_csv(train_path)
        self.test_df = pd.read_csv(test_path)
        print(f"Training data shape: {self.train_df.shape}")
        print(f"Test data shape: {self.test_df.shape}")
        return self.train_df, self.test_df

    def explore_data(self):
        """Perform basic data exploration"""
        print("\n=== DATA EXPLORATION ===")
        print("\nTraining Data Info:")
        print(self.train_df.info())
        print(f"\nMissing values in training data:\n{self.train_df.isnull().sum()}")
        print(f"\nTarget variable statistics:")
        print(self.train_df['Annual Turnover'].describe())

        # Check for outliers
        Q1 = self.train_df['Annual Turnover'].quantile(0.25)
        Q3 = self.train_df['Annual Turnover'].quantile(0.75)
        IQR = Q3 - Q1
        outliers = ((self.train_df['Annual Turnover'] < (Q1 - 1.5 * IQR)) |
                   (self.train_df['Annual Turnover'] > (Q3 + 1.5 * IQR))).sum()
        print(f"Number of outliers: {outliers}")

        # Visualize target distribution
        plt.figure(figsize=(15, 5))

        plt.subplot(1, 3, 1)
        plt.hist(self.train_df['Annual Turnover'], bins=50, edgecolor='black', alpha=0.7)
        plt.title('Distribution of Annual Turnover')
        plt.xlabel('Annual Turnover (INR)')
        plt.ylabel('Frequency')

        plt.subplot(1, 3, 2)
        plt.hist(np.log1p(self.train_df['Annual Turnover']), bins=50, edgecolor='black', alpha=0.7)
        plt.title('Log-transformed Annual Turnover')
        plt.xlabel('Log(Annual Turnover + 1)')
        plt.ylabel('Frequency')

        plt.subplot(1, 3, 3)
        plt.boxplot(self.train_df['Annual Turnover'])
        plt.title('Boxplot of Annual Turnover')
        plt.ylabel('Annual Turnover (INR)')

        plt.tight_layout()
        plt.show()

    def preprocess_data(self):
        """Enhanced data preprocessing"""
        print("\n=== DATA PREPROCESSING ===")

        # Combine train and test for consistent preprocessing
        combined_df = pd.concat([self.train_df.drop(columns=['Annual Turnover'], errors='ignore'), self.test_df], ignore_index=True, sort=False)

        # Store original target
        train_target = self.train_df['Annual Turnover'].copy()
        train_target.name = 'Annual Turnover' # Give the Series a name for the TargetEncoder


        # Handle date features
        if 'Opening Day of Restaurant' in combined_df.columns:
            combined_df['Opening Day of Restaurant'] = pd.to_datetime(
                combined_df['Opening Day of Restaurant'], errors='coerce'
            )
            # Extract useful date features
            combined_df['Opening_Year'] = combined_df['Opening Day of Restaurant'].dt.year
            combined_df['Opening_Month'] = combined_df['Opening Day of Restaurant'].dt.month
            combined_df['Opening_DayOfYear'] = combined_df['Opening Day of Restaurant'].dt.dayofyear
            combined_df['Days_Since_Opening'] = (pd.Timestamp.now() - combined_df['Opening Day of Restaurant']).dt.days

            # Create cyclical features for better representation
            combined_df['Opening_Month_Sin'] = np.sin(2 * np.pi * combined_df['Opening_Month'] / 12)
            combined_df['Opening_Month_Cos'] = np.cos(2 * np.pi * combined_df['Opening_Month'] / 12)
            combined_df['Opening_DayOfYear_Sin'] = np.sin(2 * np.pi * combined_df['Opening_DayOfYear'] / 365)
            combined_df['Opening_DayOfYear_Cos'] = np.cos(2 * np.pi * combined_df['Opening_DayOfYear'] / 365)

            combined_df = combined_df.drop('Opening Day of Restaurant', axis=1)

        # Handle missing values more intelligently
        print("Handling missing values...")

        # For numerical columns with high missing rates, create indicator variables
        high_missing_cols = ['Comedy Gigs Rating', 'Value Deals Rating', 'Live Sports Rating']
        for col in high_missing_cols:
            if col in combined_df.columns:
                combined_df[f'{col}_Missing'] = combined_df[col].isnull().astype(int)

        # Numerical columns
        numerical_cols = combined_df.select_dtypes(include=[np.number]).columns
        for col in numerical_cols:
            if col != 'Annual Turnover': # Ensure 'Annual Turnover' is not processed here
                combined_df[col] = combined_df[col].fillna(combined_df[col].median())

        # Categorical columns - Initial handling before specific encoding
        categorical_cols = combined_df.select_dtypes(include=['object']).columns
        for col in categorical_cols:
            if col != 'Registration Number':
                mode_val = combined_df[col].mode()
                if len(mode_val) > 0:
                    combined_df[col] = combined_df[col].fillna(mode_val[0])
                else:
                    combined_df[col] = combined_df[col].fillna('Unknown')

        # Apply Label Encoding to specific categorical columns BEFORE interaction terms are created
        # This ensures they are numeric for multiplication
        cols_for_label_encoding = ['Restaurant Location', 'Restaurant Type', 'EndorsedBy', 'Restaurant Theme'] # Add other low-to-medium cardinality cols
        for col in cols_for_label_encoding:
            if col in combined_df.columns:
                 le = LabelEncoder()
                 combined_df[col] = le.fit_transform(combined_df[col].astype(str))
                 self.label_encoders[col] = le # Store the encoder


        # Re-select numerical and categorical columns after preliminary encoding
        numerical_cols = combined_df.select_dtypes(include=[np.number]).columns
        categorical_cols_updated = combined_df.select_dtypes(include=['object']).columns # Should mostly be 'Registration Number' and high cardinality ones now


        # Enhanced feature engineering
        print("Creating enhanced features...")

        # Social media engagement score and ratio
        if 'Facebook Popularity Quotient' in combined_df.columns and 'Instagram Popularity Quotient' in combined_df.columns:
            combined_df['Social_Media_Score'] = (
                combined_df['Facebook Popularity Quotient'] +
                combined_df['Instagram Popularity Quotient']
            ) / 2
            # Ensure denominator is not zero; fillna after calculation
            combined_df['Social_Media_Ratio'] = combined_df['Facebook Popularity Quotient'] / (combined_df['Instagram Popularity Quotient'].replace(0, 1) + 1)
            combined_df['Social_Media_Ratio'] = combined_df['Social_Media_Ratio'].fillna(combined_df['Social_Media_Ratio'].median())


        # Service quality composite score (weighted)
        service_cols = ['Staff Responsivness', 'Hygiene Rating', 'Food Rating', 'Service']
        available_service_cols = [col for col in service_cols if col in combined_df.columns]
        if available_service_cols:
            # Weighted average (Food Rating gets higher weight)
            weights = {'Food Rating': 0.4, 'Service': 0.3, 'Hygiene Rating': 0.2, 'Staff Responsivness': 0.1}
            combined_df['Service_Quality_Score'] = 0
            total_weight = 0
            for col in available_service_cols:
                if col in combined_df.columns: # Double check column exists
                    weight = weights.get(col, 0.25)
                    combined_df['Service_Quality_Score'] += combined_df[col] * weight
                    total_weight += weight
            if total_weight > 0:
                combined_df['Service_Quality_Score'] /= total_weight
            else:
                # Handle case where none of the service columns exist
                combined_df['Service_Quality_Score'] = 0 # Or use a default value
            combined_df['Service_Quality_Score'] = combined_df['Service_Quality_Score'].fillna(combined_df['Service_Quality_Score'].median())


        # Rating consistency score
        rating_cols = ['Food Rating', 'Service', 'Hygiene Rating', 'Overall Restaurant Rating']
        available_rating_cols = [col for col in rating_cols if col in combined_df.columns]
        if len(available_rating_cols) > 1:
            combined_df['Rating_Std'] = combined_df[available_rating_cols].std(axis=1).fillna(0) # Fill NaN from std of single value
            combined_df['Rating_Range'] = combined_df[available_rating_cols].max(axis=1) - combined_df[available_rating_cols].min(axis=1)
            combined_df['Rating_Std'] = combined_df['Rating_Std'].fillna(combined_df['Rating_Std'].median())
            combined_df['Rating_Range'] = combined_df['Rating_Range'].fillna(combined_df['Rating_Range'].median())


        # Entertainment and ambience score
        entertainment_cols = ['Live Music Rating', 'Comedy Gigs Rating', 'Ambience', 'Lively']
        available_entertainment_cols = [col for col in entertainment_cols if col in combined_df.columns]
        if available_entertainment_cols:
            combined_df['Entertainment_Score'] = combined_df[available_entertainment_cols].mean(axis=1).fillna(combined_df[available_entertainment_cols].median(axis=1)) # Use median for missing
            combined_df['Entertainment_Score'] = combined_df['Entertainment_Score'].fillna(combined_df['Entertainment_Score'].median())


        # Infrastructure and convenience score
        infra_cols = ['Fire Audit', 'Liquor License Obtained', 'Situated in a Multi Complex',
                      'Dedicated Parking', 'Open Sitting Available']
        available_infra_cols = [col for col in infra_cols if col in combined_df.columns]
        if available_infra_cols:
            combined_df['Infrastructure_Score'] = combined_df[available_infra_cols].sum(axis=1)
            combined_df['Infrastructure_Score'] = combined_df['Infrastructure_Score'].fillna(combined_df['Infrastructure_Score'].median())


        # City and location features
        if 'City' in combined_df.columns:
            city_counts = combined_df['City'].value_counts()
            combined_df['City_Frequency'] = combined_df['City'].map(city_counts).fillna(0) # Fill NaN for cities only in test


        # Restaurant age feature (more sophisticated)
        if 'Days_Since_Opening' in combined_df.columns:
            combined_df['Restaurant_Age_Years'] = combined_df['Days_Since_Opening'] / 365.25
            combined_df['Is_New_Restaurant'] = (combined_df['Restaurant_Age_Years'] < 1).astype(int)
            combined_df['Is_Established'] = (combined_df['Restaurant_Age_Years'] > 5).astype(int)
            combined_df['Restaurant_Age_Years'] = combined_df['Restaurant_Age_Years'].fillna(combined_df['Restaurant_Age_Years'].median())


        # Value perception ratio
        if 'Value for Money' in combined_df.columns and 'Food Rating' in combined_df.columns:
            # Ensure denominator is not zero; fillna after calculation
            combined_df['Value_Quality_Ratio'] = combined_df['Value for Money'] / (combined_df['Food Rating'].replace(0, 1) + 1)
            combined_df['Value_Quality_Ratio'] = combined_df['Value_Quality_Ratio'].fillna(combined_df['Value_Quality_Ratio'].median())


        # *** NEW FEATURE ENGINEERING ***

        # 1. Interaction Terms (Examples based on potential domain knowledge)
        # Ensure columns are numeric before interaction - handled by preliminary LE
        if 'Social_Media_Score' in combined_df.columns and 'Restaurant Location' in combined_df.columns and pd.api.types.is_numeric_dtype(combined_df['Restaurant Location']):
             print("Creating SocialMedia_x_Location interaction...")
             combined_df['SocialMedia_x_Location'] = combined_df['Social_Media_Score'] * combined_df['Restaurant Location']
        if 'Service_Quality_Score' in combined_df.columns and 'Restaurant Type' in combined_df.columns and pd.api.types.is_numeric_dtype(combined_df['Restaurant Type']):
             print("Creating ServiceQuality_x_RestaurantType interaction...")
             combined_df['ServiceQuality_x_RestaurantType'] = combined_df['Service_Quality_Score'] * combined_df['Restaurant Type']
        if 'Infrastructure_Score' in combined_df.columns and 'Restaurant City Tier' in combined_df.columns and pd.api.types.is_numeric_dtype(combined_df['Restaurant City Tier']):
             print("Creating Infrastructure_x_CityTier interaction...")
             combined_df['Infrastructure_x_CityTier'] = combined_df['Infrastructure_Score'] * combined_df['Restaurant City Tier']
        if 'Restaurant_Age_Years' in combined_df.columns and 'Resturant Tier' in combined_df.columns and pd.api.types.is_numeric_dtype(combined_df['Resturant Tier']):
             print("Creating Age_x_Tier interaction...")
             combined_df['Age_x_Tier'] = combined_df['Restaurant_Age_Years'] * combined_df['Resturant Tier']


        # 2. Polynomial Features for highly correlated numerical features
        print("Creating polynomial features...")
        # Identify numerical columns likely to have non-linear relationships (e.g., ratings, popularity)
        # Re-include columns that might have been temporarily encoded if they were originally numeric
        polynomial_candidates = [
            'Facebook Popularity Quotient', 'Instagram Popularity Quotient',
            'Restaurant Zomato Rating', 'Value for Money', 'Food Rating',
            'Overall Restaurant Rating', 'Social_Media_Score', 'Service_Quality_Score',
            'Entertainment_Score', 'Restaurant_Age_Years', 'Restaurant City Tier', 'Resturant Tier'
        ]
        # Filter for columns actually present and numeric
        polynomial_candidates = [col for col in combined_df.columns if col in polynomial_candidates and pd.api.types.is_numeric_dtype(combined_df[col])]


        if polynomial_candidates:
            combined_df_poly = combined_df[polynomial_candidates].copy()
            # Handle potential NaNs before fitting polynomial features (using median from combined)
            combined_df_poly = combined_df_poly.fillna(combined_df_poly.median())

            # Fit on the combined data to ensure consistent transformation
            self.poly.fit(combined_df_poly)
            poly_features = self.poly.transform(combined_df_poly)

            # Create a DataFrame for polynomial features with meaningful names
            poly_feature_names = self.poly.get_feature_names_out(polynomial_candidates)
            poly_df = pd.DataFrame(poly_features, columns=poly_feature_names, index=combined_df.index)

            # Drop original polynomial candidate columns from combined_df to avoid duplication
            combined_df = combined_df.drop(columns=polynomial_candidates)

            # Concatenate the new polynomial features
            combined_df = pd.concat([combined_df, poly_df], axis=1)
            print(f"Added {poly_df.shape[1]} polynomial features.")


        # 3. More Advanced Encoding (Target Encoding for high cardinality, One-Hot for others)
        print("Applying advanced encoding...")

        # Identify categorical columns again
        categorical_cols_final = combined_df.select_dtypes(include=['object']).columns

        # Target Encode high cardinality columns on the training data first to avoid leakage
        high_cardinality_cols = []
        low_cardinality_cols = []
        for col in categorical_cols_final:
            if col not in ['Registration Number'] and col in combined_df.columns:
                unique_count = combined_df[col].nunique()
                # Use a threshold for target encoding vs one-hot, consider data size
                if unique_count > 50 and self.train_df.shape[0] > 0: # Threshold for target encoding, ensure training data exists
                    high_cardinality_cols.append(col)
                elif col != 'Registration Number':
                    low_cardinality_cols.append(col)


        # Apply Target Encoding (fit on training data, transform on combined)
        if high_cardinality_cols and self.train_df.shape[0] > 0:
            print(f"Applying Target Encoding to: {high_cardinality_cols}")
            # Split combined_df for target encoding fitting
            train_combined_part = combined_df.iloc[:len(self.train_df)].copy()

            for col in high_cardinality_cols:
                encoder = TargetEncoder(cols=[col])
                # Fit encoder on training data using the categorical column from train_combined_part and the target Series
                # This should work as temp_df in TargetEncoder is created correctly
                encoder.fit(train_combined_part, train_target) # Pass the entire DataFrame and the target Series
                self.target_encoders[col] = encoder # Store the fitted encoder

                # Transform the entire combined_df
                # Pass the entire combined_df to transform, encoder handles selecting the correct column
                combined_df[f'{col}_TargetEncoded'] = encoder.transform(combined_df)[col] # Access the transformed column name


            # Drop original high cardinality columns
            combined_df = combined_df.drop(columns=high_cardinality_cols)


        # Apply One-Hot Encoding for low cardinality columns
        if low_cardinality_cols:
            print(f"Applying One-Hot Encoding to: {low_cardinality_cols}")
            # Convert to string to avoid issues with mixed types before OHE
            for col in low_cardinality_cols:
                combined_df[col] = combined_df[col].astype(str)

            combined_df = pd.get_dummies(combined_df, columns=low_cardinality_cols, drop_first=True) # Drop first to avoid multicollinearity


        # Ensure all columns are numeric after encoding, fill any remaining NaNs
        for col in combined_df.columns:
            if combined_df[col].dtype == 'object':
                # Try to convert to numeric if possible, otherwise drop or handle specifically
                try:
                    combined_df[col] = pd.to_numeric(combined_df[col])
                except ValueError:
                    # Handle non-numeric object columns that weren't encoded
                    print(f"Warning: Column {col} remains object type after encoding. Attempting to drop.")
                    combined_df = combined_df.drop(columns=[col]) # Drop if conversion fails


        combined_df = combined_df.fillna(combined_df.median()) # Final fillna after all engineering


        # Split back into train and test
        train_size = len(self.train_df)
        self.processed_train = combined_df[:train_size].copy()
        self.processed_test = combined_df[train_size:].copy()

        # Add the original target back to the processed train df for convenience in model preparation
        self.processed_train['Annual Turnover'] = train_target


        print(f"Processed training data shape: {self.processed_train.shape}")
        print(f"Processed test data shape: {self.processed_test.shape}")

        return self.processed_train, self.processed_test


    def prepare_data(self):
        """Prepare data for models"""
        print(f"\n=== PREPARING DATA FOR MODELS ===")

        # Prepare features (exclude Registration Number and Annual Turnover from features)
        feature_cols = [col for col in self.processed_train.columns
                       if col not in ['Registration Number', 'Annual Turnover']]

        X_train = self.processed_train[feature_cols].values
        y_train = self.processed_train['Annual Turnover'].values
        X_test = self.processed_test[feature_cols].values

        print(f"Feature matrix shape: {X_train.shape}")
        print(f"Target shape: {y_train.shape}")

        # Apply log transformation to target (for scaling and potential use in models)
        y_train_log = np.log1p(y_train)

        # Scale features and target
        X_train_scaled = self.scaler_features.fit_transform(X_train)
        X_test_scaled = self.scaler_features.transform(X_test)
        y_train_scaled = self.scaler_target.fit_transform(y_train_log.reshape(-1, 1)).flatten()

        print(f"Target statistics after log+scaling - Min: {y_train_scaled.min():.3f}, Max: {y_train_scaled.max():.3f}")

        # Return original X_train, y_train for tree models that don't need scaling
        return X_train_scaled, y_train_scaled, X_test_scaled, len(feature_cols), X_train, y_train


    def create_ensemble_model(self, X_train_scaled, y_train_scaled, X_train_original, y_train_original):
        """Create ensemble with Random Forest and XGBoost"""
        print("\n=== CREATING ENSEMBLE MODEL (RF + XGBoost) ===")

        # Random Forest (uses original scale data)
        self.rf_model = RandomForestRegressor(
            n_estimators=300, # Increased estimators
            max_depth=20,     # Increased depth
            min_samples_split=5, # Reduced samples split
            min_samples_leaf=3, # Reduced samples leaf
            random_state=42,
            n_jobs=-1
        )
        self.rf_model.fit(X_train_original, y_train_original)

        # XGBoost Regressor (uses original scale data)
        self.xgb_model = xgb.XGBRegressor(
            objective='reg:squarederror',
            n_estimators=500, # Increased estimators
            learning_rate=0.03, # Reduced learning rate
            max_depth=7,        # Increased depth
            subsample=0.85,     # Slightly higher subsample
            colsample_bytree=0.85, # Slightly higher colsample
            random_state=42,
            n_jobs=-1
        )
        self.xgb_model.fit(X_train_original, y_train_original)

        # Ridge model is not part of this ensemble
        self.ridge_model = None

        return self.rf_model, self.xgb_model

    def make_ensemble_predictions(self, X_test_scaled, X_test_original):
        """Make predictions using ensemble (RF + XGBoost)"""
        # Random Forest predictions (uses original scale data)
        rf_pred = self.rf_model.predict(X_test_original)

         # XGBoost predictions (uses original scale data)
        xgb_pred = self.xgb_model.predict(X_test_original)

        # Simple averaging ensemble
        ensemble_pred = (rf_pred + xgb_pred) / 2.0


        return ensemble_pred, rf_pred, xgb_pred # Return ensemble and individual predictions

    def evaluate_model(self, X_train_scaled, y_train_scaled, X_train_original):
        """Evaluate model performance"""
        print(f"\n=== MODEL EVALUATION ===")

        # Inverse transform scaled target back to original scale
        train_actual_log = self.scaler_target.inverse_transform(y_train_scaled.reshape(-1, 1)).flatten()
        train_actual = np.expm1(train_actual_log)


        if self.use_ensemble and self.rf_model and self.xgb_model:
            # Ensemble evaluation
            train_pred_ensemble, train_pred_rf, train_pred_xgb = self.make_ensemble_predictions(
                X_train_scaled, X_train_original
            )

            # Calculate RMSE for each model
            rmse_ensemble = np.sqrt(mean_squared_error(train_actual, train_pred_ensemble))
            rmse_rf = np.sqrt(mean_squared_error(train_actual, train_pred_rf))
            rmse_xgb = np.sqrt(mean_squared_error(train_actual, train_pred_xgb))


            print(f"Ensemble (RF + XGBoost) RMSE: {rmse_ensemble:,.2f}")
            print(f"Random Forest RMSE: {rmse_rf:,.2f}")
            print(f"XGBoost RMSE: {rmse_xgb:,.2f}")


            best_rmse = rmse_ensemble
        else:
             # This case should not happen if use_ensemble is True
             print("No ensemble model available.")
             best_rmse = float('inf') # Indicate an issue


        print(f"Target RMSE: 12,240,429.84")
        print(f"Performance: {'PASSED' if best_rmse < 12240429.84 else 'NEEDS IMPROVEMENT'}")

        return best_rmse

    def create_submission_file(self, predictions, registration_numbers, filename='optimized_submission.csv'):
        """Create submission file"""
        # Ensure predictions are non-negative and round
        predictions[predictions < 0] = 0
        submission_df = pd.DataFrame({
            'Registration Number': registration_numbers,
            'Annual Turnover': np.round(predictions).astype(int)
        })

        submission_df.to_csv(filename, index=False)
        print(f"\nSubmission file saved as: {filename}")
        print(f"Submission shape: {submission_df.shape}")
        print(f"Prediction statistics:")
        print(submission_df['Annual Turnover'].describe())

        return submission_df

def main():
    """Main execution function"""
    print("=== OPTIMIZED RESTAURANT ANNUAL TURNOVER PREDICTION ===")

    # Initialize predictor
    predictor = RestaurantTurnoverPredictor()

    # File paths
    train_path = "Train_dataset_(1)_(1)_(2)_(1).csv"
    test_path = "Test_dataset_(1)_(1)_(2)_(1).csv"

    try:
        # Load data
        train_df, test_df = predictor.load_data(train_path, test_path)

        # Explore data
        predictor.explore_data()

        # Preprocess data with enhanced features
        processed_train, processed_test = predictor.preprocess_data()

        # Prepare data for models (get original and scaled data)
        X_train_scaled, y_train_scaled, X_test_scaled, n_features, X_train_original, y_train_original = predictor.prepare_data()
        # Ensure X_test_original is created correctly from processed_test
        X_test_original = processed_test[[col for col in processed_test.columns if col not in ['Registration Number', 'Annual Turnover']]].values


        # Create ensemble models (RF + XGBoost)
        if predictor.use_ensemble:
            predictor.rf_model, predictor.xgb_model = predictor.create_ensemble_model(
                X_train_scaled, y_train_scaled, X_train_original, y_train_original
            )

            # Evaluate ensemble on the training set
            train_rmse = predictor.evaluate_model(X_train_scaled, y_train_scaled, X_train_original)

            # Make ensemble predictions on the test set
            test_predictions, _, _ = predictor.make_ensemble_predictions(X_test_scaled, X_test_original)
        else:
             print("Ensemble is not enabled.")
             test_predictions = np.zeros(len(processed_test)) # Placeholder


        # Create submission file
        submission_df = predictor.create_submission_file(
            test_predictions,
            processed_test['Registration Number'].values
        )

        print(f"\n=== PIPELINE COMPLETED SUCCESSFULLY ===")
        print(f"Final Training RMSE: {train_rmse:,.2f}")

        return predictor # Return the predictor object

    except FileNotFoundError:
        print("\nERROR: Data files not found!")
        print("Please ensure the following files are in your working directory:")
        print("- Train_dataset_(1)_(1)_(2)_(1).csv")
        return None # Return None on error
    except Exception as e:
        print(f"\nERROR: {str(e)}")
        import traceback
        traceback.print_exc()
        return None # Return None on error

# Execute the main function to run the pipeline with enhanced features
predictor = main()
```

```python
class RestaurantTurnoverPredictor:
    def __init__(self):
        self.scaler_features = RobustScaler()
        self.scaler_target = RobustScaler()
        self.label_encoders = {}
        self.target_encoders = {}
        self.rf_model = None
        self.ridge_model = None
        self.xgb_model = None
        self.use_ensemble = True
        self.poly = PolynomialFeatures(degree=2, include_bias=False)
        # Store capping values learned from training data
        self.target_cap_value = None
        self.feature_caps = {}


    def load_data(self, train_path, test_path):
        """Load training and test datasets"""
        self.train_df = pd.read_csv(train_path)
        self.test_df = pd.read_csv(test_path)
        print(f"Training data shape: {self.train_df.shape}")
        print(f"Test data shape: {self.test_df.shape}")
        return self.train_df, self.test_df

    def explore_data(self):
        """Perform basic data exploration"""
        print("\n=== DATA EXPLORATION ===")
        print("\nTraining Data Info:")
        print(self.train_df.info())
        print(f"\nMissing values in training data:\n{self.train_df.isnull().sum()}")
        print(f"\nTarget variable statistics:")
        print(self.train_df['Annual Turnover'].describe())

        # Check for outliers
        Q1 = self.train_df['Annual Turnover'].quantile(0.25)
        Q3 = self.train_df['Annual Turnover'].quantile(0.75)
        IQR = Q3 - Q1
        outliers = ((self.train_df['Annual Turnover'] < (Q1 - 1.5 * IQR)) |
                   (self.train_df['Annual Turnover'] > (Q3 + 1.5 * IQR))).sum()
        print(f"Number of outliers: {outliers}")

        # Visualize target distribution
        plt.figure(figsize=(15, 5))

        plt.subplot(1, 3, 1)
        plt.hist(self.train_df['Annual Turnover'], bins=50, edgecolor='black', alpha=0.7)
        plt.title('Distribution of Annual Turnover')
        plt.xlabel('Annual Turnover (INR)')
        plt.ylabel('Frequency')

        plt.subplot(1, 3, 2)
        plt.hist(np.log1p(self.train_df['Annual Turnover']), bins=50, edgecolor='black', alpha=0.7)
        plt.title('Log-transformed Annual Turnover')
        plt.xlabel('Log(Annual Turnover + 1)')
        plt.ylabel('Frequency')

        plt.subplot(1, 3, 3)
        plt.boxplot(self.train_df['Annual Turnover'])
        plt.title('Boxplot of Annual Turnover')
        plt.ylabel('Annual Turnover (INR)')

        plt.tight_layout()
        plt.show()

    def preprocess_data(self):
        """Enhanced data preprocessing with outlier treatment"""
        print("\n=== DATA PREPROCESSING ===")

        # Handle target variable outliers BEFORE combining
        print("Handling target outliers...")
        # Calculate 99th percentile cap from training data
        self.target_cap_value = self.train_df['Annual Turnover'].quantile(0.99)
        # Apply capping ONLY to the training data
        self.train_df['Annual Turnover_Capped'] = self.train_df['Annual Turnover'].clip(upper=self.target_cap_value)
        print(f"Target variable capped at {self.target_cap_value:,.2f} (99th percentile).")
        print(f"Original max: {self.train_df['Annual Turnover'].max():,.2f}, Capped max: {self.train_df['Annual Turnover_Capped'].max():,.2f}")


        # Combine train and test for consistent feature preprocessing
        # Use the capped target from train_df
        combined_df = pd.concat([self.train_df.drop(columns=['Annual Turnover'], errors='ignore').rename(columns={'Annual Turnover_Capped': 'Annual Turnover'}),
                                self.test_df], ignore_index=True, sort=False)

        # Store original target (capped) for Target Encoding and model training
        train_target = self.train_df['Annual Turnover_Capped'].copy()
        train_target.name = 'Annual Turnover'


        # Handle date features
        if 'Opening Day of Restaurant' in combined_df.columns:
            combined_df['Opening Day of Restaurant'] = pd.to_datetime(
                combined_df['Opening Day of Restaurant'], errors='coerce'
            )
            # Extract useful date features
            combined_df['Opening_Year'] = combined_df['Opening Day of Restaurant'].dt.year
            combined_df['Opening_Month'] = combined_df['Opening Day of Restaurant'].dt.month
            combined_df['Opening_DayOfYear'] = combined_df['Opening Day of Restaurant'].dt.dayofyear
            combined_df['Days_Since_Opening'] = (pd.Timestamp.now() - combined_df['Opening Day of Restaurant']).dt.days

            # Create cyclical features for better representation
            combined_df['Opening_Month_Sin'] = np.sin(2 * np.pi * combined_df['Opening_Month'] / 12)
            combined_df['Opening_Month_Cos'] = np.cos(2 * np.pi * combined_df['Opening_Month'] / 12)
            combined_df['Opening_DayOfYear_Sin'] = np.sin(2 * np.pi * combined_df['Opening_DayOfYear'] / 365)
            combined_df['Opening_DayOfYear_Cos'] = np.cos(2 * np.pi * combined_df['Opening_DayOfYear'] / 365)

            combined_df = combined_df.drop('Opening Day of Restaurant', axis=1)

        # Handle missing values more intelligently
        print("Handling missing values...")

        # For numerical columns with high missing rates, create indicator variables
        high_missing_cols = ['Comedy Gigs Rating', 'Value Deals Rating', 'Live Sports Rating']
        for col in high_missing_cols:
            if col in combined_df.columns:
                combined_df[f'{col}_Missing'] = combined_df[col].isnull().astype(int)

        # Categorical columns - Initial handling before specific encoding
        categorical_cols = combined_df.select_dtypes(include=['object']).columns
        for col in categorical_cols:
            if col != 'Registration Number':
                mode_val = combined_df[col].mode()
                if len(mode_val) > 0:
                    combined_df[col] = combined_df[col].fillna(mode_val[0])
                else:
                    combined_df[col] = combined_df[col].fillna('Unknown')

        # Apply Label Encoding to specific categorical columns BEFORE interaction terms are created
        # This ensures they are numeric for multiplication
        cols_for_label_encoding = ['Restaurant Location', 'Restaurant Type', 'EndorsedBy', 'Restaurant Theme'] # Add other low-to-medium cardinality cols
        for col in cols_for_label_encoding:
            if col in combined_df.columns:
                 le = LabelEncoder()
                 combined_df[col] = le.fit_transform(combined_df[col].astype(str))
                 self.label_encoders[col] = le # Store the encoder


        # Re-select numerical columns after preliminary encoding
        numerical_cols = combined_df.select_dtypes(include=[np.number]).columns


        # Handle numerical feature outliers AFTER initial imputation but BEFORE scaling/polynomial features
        print("Handling numerical feature outliers...")
        # Select features based on prior outlier analysis (e.g., top N or those with > X% outliers)
        # Using a threshold of > 5% outliers as a starting point
        features_to_cap = [col for col, count in outlier_counts.items() if count / self.train_df.shape[0] > 0.05 and col in numerical_cols]
        print(f"Applying capping to {len(features_to_cap)} features: {features_to_cap}")

        for col in features_to_cap:
            # Calculate 99th percentile cap from the TRAINING PORTION of the combined data
            train_portion = combined_df[:len(self.train_df)][col]
            cap_value = train_portion.quantile(0.99)
            self.feature_caps[col] = cap_value # Store the cap value
            # Apply capping to the entire combined DataFrame
            combined_df[col] = combined_df[col].clip(upper=cap_value)
            print(f"  - {col} capped at {cap_value:,.2f}")


        # Enhanced feature engineering (re-ordered to come after outlier treatment)
        print("Creating enhanced features...")

        # Social media engagement score and ratio
        if 'Facebook Popularity Quotient' in combined_df.columns and 'Instagram Popularity Quotient' in combined_df.columns:
            combined_df['Social_Media_Score'] = (
                combined_df['Facebook Popularity Quotient'] +
                combined_df['Instagram Popularity Quotient']
            ) / 2
            # Ensure denominator is not zero; fillna after calculation
            combined_df['Social_Media_Ratio'] = combined_df['Facebook Popularity Quotient'] / (combined_df['Instagram Popularity Quotient'].replace(0, 1) + 1)
            combined_df['Social_Media_Ratio'] = combined_df['Social_Media_Ratio'].fillna(combined_df['Social_Media_Ratio'].median())


        # Service quality composite score (weighted)
        service_cols = ['Staff Responsivness', 'Hygiene Rating', 'Food Rating', 'Service']
        available_service_cols = [col for col in service_cols if col in combined_df.columns]
        if available_service_cols:
            # Weighted average (Food Rating gets higher weight)
            weights = {'Food Rating': 0.4, 'Service': 0.3, 'Hygiene Rating': 0.2, 'Staff Responsivness': 0.1}
            combined_df['Service_Quality_Score'] = 0
            total_weight = 0
            for col in available_service_cols:
                if col in combined_df.columns: # Double check column exists
                    weight = weights.get(col, 0.25)
                    combined_df['Service_Quality_Score'] += combined_df[col] * weight
                    total_weight += weight
            if total_weight > 0:
                combined_df['Service_Quality_Score'] /= total_weight
            else:
                # Handle case where none of the service columns exist
                combined_df['Service_Quality_Score'] = 0 # Or use a default value
            combined_df['Service_Quality_Score'] = combined_df['Service_Quality_Score'].fillna(combined_df['Service_Quality_Score'].median())


        # Rating consistency score
        rating_cols = ['Food Rating', 'Service', 'Hygiene Rating', 'Overall Restaurant Rating']
        available_rating_cols = [col for col in rating_cols if col in combined_df.columns]
        if len(available_rating_cols) > 1:
            combined_df['Rating_Std'] = combined_df[available_rating_cols].std(axis=1).fillna(0) # Fill NaN from std of single value
            combined_df['Rating_Range'] = combined_df[available_rating_cols].max(axis=1) - combined_df[available_rating_cols].min(axis=1)
            combined_df['Rating_Std'] = combined_df['Rating_Std'].fillna(combined_df['Rating_Std'].median())
            combined_df['Rating_Range'] = combined_df['Rating_Range'].fillna(combined_df['Rating_Range'].median())


        # Entertainment and ambience score
        entertainment_cols = ['Live Music Rating', 'Comedy Gigs Rating', 'Ambience', 'Lively']
        available_entertainment_cols = [col for col in entertainment_cols if col in combined_df.columns]
        if available_entertainment_cols:
            combined_df['Entertainment_Score'] = combined_df[available_entertainment_cols].mean(axis=1).fillna(combined_df[available_entertainment_cols].median(axis=1)) # Use median for missing
            combined_df['Entertainment_Score'] = combined_df['Entertainment_Score'].fillna(combined_df['Entertainment_Score'].median())


        # Infrastructure and convenience score
        infra_cols = ['Fire Audit', 'Liquor License Obtained', 'Situated in a Multi Complex',
                      'Dedicated Parking', 'Open Sitting Available']
        available_infra_cols = [col for col in infra_cols if col in combined_df.columns]
        if available_infra_cols:
            combined_df['Infrastructure_Score'] = combined_df[available_infra_cols].sum(axis=1)
            combined_df['Infrastructure_Score'] = combined_df['Infrastructure_Score'].fillna(combined_df['Infrastructure_Score'].median())


        # City and location features
        if 'City' in combined_df.columns:
            city_counts = combined_df['City'].value_counts()
            combined_df['City_Frequency'] = combined_df['City'].map(city_counts).fillna(0) # Fill NaN for cities only in test


        # Restaurant age feature (more sophisticated)
        if 'Days_Since_Opening' in combined_df.columns:
            combined_df['Restaurant_Age_Years'] = combined_df['Days_Since_Opening'] / 365.25
            combined_df['Is_New_Restaurant'] = (combined_df['Restaurant_Age_Years'] < 1).astype(int)
            combined_df['Is_Established'] = (combined_df['Restaurant_Age_Years'] > 5).astype(int)
            combined_df['Restaurant_Age_Years'] = combined_df['Restaurant_Age_Years'].fillna(combined_df['Restaurant_Age_Years'].median())


        # Value perception ratio
        if 'Value for Money' in combined_df.columns and 'Food Rating' in combined_df.columns:
            # Ensure denominator is not zero; fillna after calculation
            combined_df['Value_Quality_Ratio'] = combined_df['Value for Money'] / (combined_df['Food Rating'].replace(0, 1) + 1)
            combined_df['Value_Quality_Ratio'] = combined_df['Value_Quality_Ratio'].fillna(combined_df['Value_Quality_Ratio'].median())


        # *** Interaction Terms ***
        # Ensure columns are numeric before interaction - handled by preliminary LE
        if 'Social_Media_Score' in combined_df.columns and 'Restaurant Location' in combined_df.columns and pd.api.types.is_numeric_dtype(combined_df['Restaurant Location']):
             print("Creating SocialMedia_x_Location interaction...")
             combined_df['SocialMedia_x_Location'] = combined_df['Social_Media_Score'] * combined_df['Restaurant Location']
        if 'Service_Quality_Score' in combined_df.columns and 'Restaurant Type' in combined_df.columns and pd.api.types.is_numeric_dtype(combined_df['Restaurant Type']):
             print("Creating ServiceQuality_x_RestaurantType interaction...")
             combined_df['ServiceQuality_x_RestaurantType'] = combined_df['Service_Quality_Score'] * combined_df['Restaurant Type']
        if 'Infrastructure_Score' in combined_df.columns and 'Restaurant City Tier' in combined_df.columns and pd.api.types.is_numeric_dtype(combined_df['Restaurant City Tier']):
             print("Creating Infrastructure_x_CityTier interaction...")
             combined_df['Infrastructure_x_CityTier'] = combined_df['Infrastructure_Score'] * combined_df['Restaurant City Tier']
        if 'Restaurant_Age_Years' in combined_df.columns and 'Resturant Tier' in combined_df.columns and pd.api.types.is_numeric_dtype(combined_df['Resturant Tier']):
             print("Creating Age_x_Tier interaction...")
             combined_df['Age_x_Tier'] = combined_df['Restaurant_Age_Years'] * combined_df['Resturant Tier']


        # *** Polynomial Features ***
        print("Creating polynomial features...")
        polynomial_candidates = [
            'Facebook Popularity Quotient', 'Instagram Popularity Quotient',
            'Restaurant Zomato Rating', 'Value for Money', 'Food Rating',
            'Overall Restaurant Rating', 'Social_Media_Score', 'Service_Quality_Score',
            'Entertainment_Score', 'Restaurant_Age_Years', 'Restaurant City Tier', 'Resturant Tier'
        ]
        polynomial_candidates = [col for col in combined_df.columns if col in polynomial_candidates and pd.api.types.is_numeric_dtype(combined_df[col])]


        if polynomial_candidates:
            combined_df_poly = combined_df[polynomial_candidates].copy()
            # Handle potential NaNs before fitting polynomial features (using median from combined)
            combined_df_poly = combined_df_poly.fillna(combined_df_poly.median())

            # Fit on the combined data to ensure consistent transformation
            self.poly.fit(combined_df_poly)
            poly_features = self.poly.transform(combined_df_poly)

            # Create a DataFrame for polynomial features with meaningful names
            poly_feature_names = self.poly.get_feature_names_out(polynomial_candidates)
            poly_df = pd.DataFrame(poly_features, columns=poly_feature_names, index=combined_df.index)

            # Drop original polynomial candidate columns from combined_df to avoid duplication
            combined_df = combined_df.drop(columns=polynomial_candidates)

            # Concatenate the new polynomial features
            combined_df = pd.concat([combined_df, poly_df], axis=1)
            print(f"Added {poly_df.shape[1]} polynomial features.")


        # *** More Advanced Encoding (Target Encoding for high cardinality, One-Hot for others) ***
        print("Applying advanced encoding...")

        # Identify categorical columns again
        categorical_cols_final = combined_df.select_dtypes(include=['object']).columns

        # Target Encode high cardinality columns on the training data first to avoid leakage
        high_cardinality_cols = []
        low_cardinality_cols = []
        for col in categorical_cols_final:
            if col not in ['Registration Number'] and col in combined_df.columns:
                unique_count = combined_df[col].nunique()
                # Use a threshold for target encoding vs one-hot, consider data size
                if unique_count > 50 and self.train_df.shape[0] > 0: # Threshold for target encoding, ensure training data exists
                    high_cardinality_cols.append(col)
                elif col != 'Registration Number':
                    low_cardinality_cols.append(col)


        # Apply Target Encoding (fit on training data, transform on combined)
        if high_cardinality_cols and self.train_df.shape[0] > 0:
            print(f"Applying Target Encoding to: {high_cardinality_cols}")
            # Split combined_df for target encoding fitting
            train_combined_part = combined_df[:len(self.train_df)].copy()

            for col in high_cardinality_cols:
                encoder = TargetEncoder(cols=[col])
                # Fit encoder on training data using the categorical column from train_combined_part and the target Series
                # This should work as temp_df in TargetEncoder is created correctly
                encoder.fit(train_combined_part, train_target) # Pass the entire DataFrame and the target Series
                self.target_encoders[col] = encoder # Store the fitted encoder

                # Transform the entire combined_df
                # Pass the entire combined_df to transform, encoder handles selecting the correct column
                combined_df[f'{col}_TargetEncoded'] = encoder.transform(combined_df)[col] # Access the transformed column name


            # Drop original high cardinality columns
            combined_df = combined_df.drop(columns=high_cardinality_cols)


        # Apply One-Hot Encoding for low cardinality columns
        if low_cardinality_cols:
            print(f"Applying One-Hot Encoding to: {low_cardinality_cols}")
            # Convert to string to avoid issues with mixed types before OHE
            for col in low_cardinality_cols:
                combined_df[col] = combined_df[col].astype(str)

            combined_df = pd.get_dummies(combined_df, columns=low_cardinality_cols, drop_first=True) # Drop first to avoid multicollinearity


        # Ensure all columns are numeric after encoding, fill any remaining NaNs
        for col in combined_df.columns:
            if combined_df[col].dtype == 'object':
                # Try to convert to numeric if possible, otherwise drop or handle specifically
                try:
                    combined_df[col] = pd.to_numeric(combined_df[col])
                except ValueError:
                    # Handle non-numeric object columns that weren't encoded
                    print(f"Warning: Column {col} remains object type after encoding. Attempting to drop.")
                    combined_df = combined_df.drop(columns=[col]) # Drop if conversion fails


        combined_df = combined_df.fillna(combined_df.median()) # Final fillna after all engineering


        # Split back into train and test
        train_size = len(self.train_df)
        self.processed_train = combined_df[:train_size].copy()
        self.processed_test = combined_df[train_size:].copy()

        # Add the original target back to the processed train df for convenience in model preparation
        # Note: This is the CAPPED target
        self.processed_train['Annual Turnover'] = train_target


        print(f"Processed training data shape: {self.processed_train.shape}")
        print(f"Processed test data shape: {self.processed_test.shape}")

        return self.processed_train, self.processed_test


    def prepare_data(self):
        """Prepare data for models"""
        print(f"\n=== PREPARING DATA FOR MODELS ===")

        # Prepare features (exclude Registration Number and Annual Turnover from features)
        feature_cols = [col for col in self.processed_train.columns
                       if col not in ['Registration Number', 'Annual Turnover']]

        X_train = self.processed_train[feature_cols].values
        # Use the CAPPED target for training
        y_train = self.processed_train['Annual Turnover'].values
        X_test = self.processed_test[feature_cols].values

        print(f"Feature matrix shape: {X_train.shape}")
        print(f"Target shape: {y_train.shape}")

        # Apply log transformation to target (for scaling and potential use in models)
        y_train_log = np.log1p(y_train)

        # Scale features and target
        X_train_scaled = self.scaler_features.fit_transform(X_train)
        X_test_scaled = self.scaler_features.transform(X_test)
        y_train_scaled = self.scaler_target.fit_transform(y_train_log.reshape(-1, 1)).flatten()

        print(f"Target statistics after log+scaling - Min: {y_train_scaled.min():.3f}, Max: {y_train_scaled.max():.3f}")

        # Return original X_train, y_train (capped) for tree models that don't need scaling
        return X_train_scaled, y_train_scaled, X_test_scaled, len(feature_cols), X_train, y_train


    def create_ensemble_model(self, X_train_scaled, y_train_scaled, X_train_original, y_train_original):
        """Create ensemble with Random Forest and XGBoost"""
        print("\n=== CREATING ENSEMBLE MODEL (RF + XGBoost) ===")

        # Random Forest (uses original scale data)
        self.rf_model = RandomForestRegressor(
            n_estimators=300, # Increased estimators
            max_depth=20,     # Increased depth
            min_samples_split=5, # Reduced samples split
            min_samples_leaf=3, # Reduced samples leaf
            random_state=42,
            n_jobs=-1
        )
        self.rf_model.fit(X_train_original, y_train_original)

        # XGBoost Regressor (uses original scale data)
        self.xgb_model = xgb.XGBRegressor(
            objective='reg:squarederror',
            n_estimators=500, # Increased estimators
            learning_rate=0.03, # Reduced learning rate
            max_depth=7,        # Increased depth
            subsample=0.85,     # Slightly higher subsample
            colsample_bytree=0.85, # Slightly higher colsample
            random_state=42,
            n_jobs=-1
        )
        self.xgb_model.fit(X_train_original, y_train_original)

        # Ridge model is not part of this ensemble
        self.ridge_model = None

        return self.rf_model, self.xgb_model

    def make_ensemble_predictions(self, X_test_scaled, X_test_original):
        """Make predictions using ensemble (RF + XGBoost)"""
        # Random Forest predictions (uses original scale data)
        rf_pred = self.rf_model.predict(X_test_original)

         # XGBoost predictions (uses original scale data)
        xgb_pred = self.xgb_model.predict(X_test_original)

        # Simple averaging ensemble
        ensemble_pred = (rf_pred + xgb_pred) / 2.0

        # Inverse transform predictions from the capped target scale back to original
        # Note: The models were trained on the capped target, so their predictions are on the capped scale.
        # We need to inverse transform from the scaled-log-capped space.
        # However, tree models (RF, XGBoost) are trained on the original (but capped) y,
        # so their predictions are directly on the capped scale.
        # The previous make_predictions method handled inverse scaling from the NN's output.
        # For tree models, we just need to make sure the predictions are non-negative.
        ensemble_pred = np.maximum(0, ensemble_pred)
        rf_pred = np.maximum(0, rf_pred)
        xgb_pred = np.maximum(0, xgb_pred)


        return ensemble_pred, rf_pred, xgb_pred # Return ensemble and individual predictions

    def evaluate_model(self, X_train_scaled, y_train_scaled, X_train_original):
        """Evaluate model performance"""
        print(f"\n=== MODEL EVALUATION ===")

        # Get the original, uncapped target for evaluation
        train_actual = self.train_df['Annual Turnover'].values

        if self.use_ensemble and self.rf_model and self.xgb_model:
            # Ensemble evaluation
            train_pred_ensemble, train_pred_rf, train_pred_xgb = self.make_ensemble_predictions(
                X_train_scaled, X_train_original
            )

            # Calculate RMSE using the ORIGINAL (uncapped) target
            rmse_ensemble = np.sqrt(mean_squared_error(train_actual, train_pred_ensemble))
            rmse_rf = np.sqrt(mean_squared_error(train_actual, train_pred_rf))
            rmse_xgb = np.sqrt(mean_squared_error(train_actual, train_pred_xgb))


            print(f"Ensemble (RF + XGBoost) RMSE (on original target): {rmse_ensemble:,.2f}")
            print(f"Random Forest RMSE (on original target): {rmse_rf:,.2f}")
            print(f"XGBoost RMSE (on original target): {rmse_xgb:,.2f}")


            best_rmse = rmse_ensemble
        else:
             # This case should not happen if use_ensemble is True
             print("No ensemble model available.")
             best_rmse = float('inf') # Indicate an issue


        print(f"Target RMSE: 12,240,429.84")
        print(f"Performance: {'PASSED' if best_rmse < 12240429.84 else 'NEEDS IMPROVEMENT'}")

        return best_rmse

    def create_submission_file(self, predictions, registration_numbers, filename='optimized_submission.csv'):
        """Create submission file"""
        # Ensure predictions are non-negative and round
        predictions[predictions < 0] = 0
        submission_df = pd.DataFrame({
            'Registration Number': registration_numbers,
            'Annual Turnover': np.round(predictions).astype(int)
        })

        submission_df.to_csv(filename, index=False)
        print(f"\nSubmission file saved as: {filename}")
        print(f"Submission shape: {submission_df.shape}")
        print(f"Prediction statistics:")
        print(submission_df['Annual Turnover'].describe())

        return submission_df

def main():
    """Main execution function"""
    print("=== OPTIMIZED RESTAURANT ANNUAL TURNOVER PREDICTION ===")

    # Initialize predictor
    predictor = RestaurantTurnoverPredictor()

    # File paths
    train_path = "Train_dataset_(1)_(1)_(2)_(1).csv"
    test_path = "Test_dataset_(1)_(1)_(2)_(1).csv"

    try:
        # Load data
        train_df, test_df = predictor.load_data(train_path, test_path)

        # Explore data
        predictor.explore_data()

        # Preprocess data with enhanced features and outlier treatment
        processed_train, processed_test = predictor.preprocess_data()

        # Prepare data for models (get original and scaled data)
        X_train_scaled, y_train_scaled, X_test_scaled, n_features, X_train_original, y_train_original = predictor.prepare_data()
        # Ensure X_test_original is created correctly from processed_test
        X_test_original = processed_test[[col for col in processed_test.columns if col not in ['Registration Number', 'Annual Turnover']]].values


        # Create ensemble models (RF + XGBoost)
        if predictor.use_ensemble:
            predictor.rf_model, predictor.xgb_model = predictor.create_ensemble_model(
                X_train_scaled, y_train_scaled, X_train_original, y_train_original
            )

            # Evaluate ensemble on the training set (using original, uncapped target)
            train_rmse = predictor.evaluate_model(X_train_scaled, y_train_scaled, X_train_original)

            # Make ensemble predictions on the test set
            test_predictions, _, _ = predictor.make_ensemble_predictions(X_test_scaled, X_test_original)
        else:
             print("Ensemble is not enabled.")
             test_predictions = np.zeros(len(processed_test)) # Placeholder


        # Create submission file
        submission_df = predictor.create_submission_file(
            test_predictions,
            processed_test['Registration Number'].values
        )

        print(f"\n=== PIPELINE COMPLETED SUCCESSFULLY ===")
        print(f"Final Training RMSE: {train_rmse:,.2f}")

        return predictor # Return the predictor object

    except FileNotFoundError:
        print("\nERROR: Data files not found!")
        print("Please ensure the following files are in your working directory:")
        print("- Train_dataset_(1)_(1)_(2)_(1).csv")
        return None # Return None on error
    except Exception as e:
        print(f"\nERROR: {str(e)}")
        import traceback
        traceback.print_exc()
        return None # Return None on error

# Execute the main function to run the pipeline with outlier treatment
predictor = main()
```

```python
class RestaurantTurnoverPredictor:
    def __init__(self):
        self.scaler_features = RobustScaler()
        self.scaler_target = RobustScaler()
        self.label_encoders = {}
        self.target_encoders = {}
        self.rf_model = None
        self.xgb_model = None
        self.use_ensemble = True
        self.poly = PolynomialFeatures(degree=2, include_bias=False)
        # Store capping values learned from training data (keeping for potential future use)
        self.target_cap_value = None
        self.feature_caps = {}
        # Store median values for imputation (for numerical features)
        self.numerical_medians = {}


    def load_data(self, train_path, test_path):
        """Load training and test datasets"""
        self.train_df = pd.read_csv(train_path)
        self.test_df = pd.read_csv(test_path)
        print(f"Training data shape: {self.train_df.shape}")
        print(f"Test data shape: {self.test_df.shape}")
        return self.train_df, self.test_df

    def explore_data(self):
        """Perform basic data exploration"""
        print("\n=== DATA EXPLORATION ===")
        print("\nTraining Data Info:")
        print(self.train_df.info())
        print(f"\nMissing values in training data:\n{self.train_df.isnull().sum()}")
        print(f"\nTarget variable statistics:")
        print(self.train_df['Annual Turnover'].describe())

        # Check for outliers
        Q1 = self.train_df['Annual Turnover'].quantile(0.25)
        Q3 = self.train_df['Annual Turnover'].quantile(0.75)
        IQR = Q3 - Q1
        outliers = ((self.train_df['Annual Turnover'] < (Q1 - 1.5 * IQR)) |
                   (self.train_df['Annual Turnover'] > (Q3 + 1.5 * IQR))).sum()
        print(f"Number of outliers: {outliers}")

        # Visualize target distribution
        plt.figure(figsize=(15, 5))

        plt.subplot(1, 3, 1)
        plt.hist(self.train_df['Annual Turnover'], bins=50, edgecolor='black', alpha=0.7)
        plt.title('Distribution of Annual Turnover')
        plt.xlabel('Annual Turnover (INR)')
        plt.ylabel('Frequency')

        plt.subplot(1, 3, 2)
        plt.hist(np.log1p(self.train_df['Annual Turnover']), bins=50, edgecolor='black', alpha=0.7)
        plt.title('Log-transformed Annual Turnover')
        plt.xlabel('Log(Annual Turnover + 1)')
        plt.ylabel('Frequency')

        plt.subplot(1, 3, 3)
        plt.boxplot(self.train_df['Annual Turnover'])
        plt.title('Boxplot of Annual Turnover')
        plt.ylabel('Annual Turnover (INR)')

        plt.tight_layout()
        plt.show()

    def preprocess_data(self):
        """Enhanced data preprocessing with alternative outlier/missing treatment"""
        print("\n=== DATA PREPROCESSING ===")

        # Keep original target for evaluation later
        original_train_target = self.train_df['Annual Turnover'].copy()

        # Combine train and test for consistent feature preprocessing
        # Drop the original target column as we'll use the capped/transformed one for training if needed
        combined_df = pd.concat([self.train_df.drop(columns=['Annual Turnover'], errors='ignore'),
                                self.test_df], ignore_index=True, sort=False)

        # Handle date features
        if 'Opening Day of Restaurant' in combined_df.columns:
            combined_df['Opening Day of Restaurant'] = pd.to_datetime(
                combined_df['Opening Day of Restaurant'], errors='coerce'
            )
            # Extract useful date features
            combined_df['Opening_Year'] = combined_df['Opening Day of Restaurant'].dt.year
            combined_df['Opening_Month'] = combined_df['Opening Day of Restaurant'].dt.month
            combined_df['Opening_DayOfYear'] = combined_df['Opening Day of Restaurant'].dt.dayofyear
            combined_df['Days_Since_Opening'] = (pd.Timestamp.now() - combined_df['Opening Day of Restaurant']).dt.days

            # Create cyclical features for better representation
            combined_df['Opening_Month_Sin'] = np.sin(2 * np.pi * combined_df['Opening_Month'] / 12)
            combined_df['Opening_Month_Cos'] = np.cos(2 * np.pi * combined_df['Opening_Month'] / 12)
            combined_df['Opening_DayOfYear_Sin'] = np.sin(2 * np.pi * combined_df['Opening_DayOfYear'] / 365)
            combined_df['Opening_DayOfYear_Cos'] = np.cos(2 * np.pi * combined_df['Opening_DayOfYear'] / 365)

            combined_df = combined_df.drop('Opening Day of Restaurant', axis=1)

        # Handle missing values more intelligently
        print("Handling missing values...")

        # For numerical columns with high missing rates, create indicator variables
        high_missing_cols = ['Comedy Gigs Rating', 'Value Deals Rating', 'Live Sports Rating']
        for col in high_missing_cols:
            if col in combined_df.columns:
                combined_df[f'{col}_Missing'] = combined_df[col].isnull().astype(int)

        # Numerical columns - Impute with median (learned from training data)
        numerical_cols = combined_df.select_dtypes(include=[np.number]).columns
        # Calculate medians from the training portion only
        train_numerical_data = combined_df[:len(self.train_df)][numerical_cols]
        self.numerical_medians = train_numerical_data.median().to_dict()

        for col in numerical_cols:
            combined_df[col] = combined_df[col].fillna(self.numerical_medians.get(col, combined_df[col].median())) # Fallback to overall median if col not in train_numerical_data


        # Categorical columns - Impute with mode (learned from training data)
        categorical_cols = combined_df.select_dtypes(include=['object']).columns
        # Calculate modes from the training portion only
        train_categorical_data = combined_df[:len(self.train_df)][categorical_cols]
        categorical_modes = {col: train_categorical_data[col].mode()[0] if not train_categorical_data[col].mode().empty else 'Unknown' for col in categorical_cols if col != 'Registration Number'}

        for col in categorical_cols:
            if col != 'Registration Number':
                combined_df[col] = combined_df[col].fillna(categorical_modes.get(col, 'Unknown'))


        # Apply Label Encoding to specific categorical columns BEFORE interaction terms are created
        cols_for_label_encoding = ['Restaurant Location', 'Restaurant Type', 'EndorsedBy', 'Restaurant Theme']
        for col in cols_for_label_encoding:
            if col in combined_df.columns:
                 le = LabelEncoder()
                 # Fit encoder on the combined data to handle all categories
                 combined_df[col] = le.fit_transform(combined_df[col].astype(str))
                 self.label_encoders[col] = le


        # Re-select numerical columns after preliminary encoding
        numerical_cols = combined_df.select_dtypes(include=[np.number]).columns


        # Alternative Outlier Treatment for features: Do NOT cap, rely on RobustScaler and tree models
        # The previous capping might have been too aggressive. RobustScaler handles outliers well.
        # Tree models (RF, XGBoost) are less sensitive to outliers than linear models or NNs.
        print("Skipping numerical feature outlier capping. Relying on RobustScaler and tree models.")


        # Enhanced feature engineering (order matters - after imputation, before scaling)
        print("Creating enhanced features...")

        # Social media engagement score and ratio
        if 'Facebook Popularity Quotient' in combined_df.columns and 'Instagram Popularity Quotient' in combined_df.columns:
            combined_df['Social_Media_Score'] = (
                combined_df['Facebook Popularity Quotient'] +
                combined_df['Instagram Popularity Quotient']
            ) / 2
            combined_df['Social_Media_Ratio'] = combined_df['Facebook Popularity Quotient'] / (combined_df['Instagram Popularity Quotient'].replace(0, 1) + 1)
            combined_df['Social_Media_Ratio'] = combined_df['Social_Media_Ratio'].fillna(combined_df['Social_Media_Ratio'].median())


        # Service quality composite score (weighted)
        service_cols = ['Staff Responsivness', 'Hygiene Rating', 'Food Rating', 'Service']
        available_service_cols = [col for col in service_cols if col in combined_df.columns]
        if available_service_cols:
            weights = {'Food Rating': 0.4, 'Service': 0.3, 'Hygiene Rating': 0.2, 'Staff Responsivness': 0.1}
            combined_df['Service_Quality_Score'] = 0
            total_weight = 0
            for col in available_service_cols:
                if col in combined_df.columns:
                    weight = weights.get(col, 0.25)
                    combined_df['Service_Quality_Score'] += combined_df[col] * weight
                    total_weight += weight
            if total_weight > 0:
                combined_df['Service_Quality_Score'] /= total_weight
            else:
                combined_df['Service_Quality_Score'] = 0
            combined_df['Service_Quality_Score'] = combined_df['Service_Quality_Score'].fillna(combined_df['Service_Quality_Score'].median())


        # Rating consistency score
        rating_cols = ['Food Rating', 'Service', 'Hygiene Rating', 'Overall Restaurant Rating']
        available_rating_cols = [col for col in rating_cols if col in combined_df.columns]
        if len(available_rating_cols) > 1:
            combined_df['Rating_Std'] = combined_df[available_rating_cols].std(axis=1).fillna(0)
            combined_df['Rating_Range'] = combined_df[available_rating_cols].max(axis=1) - combined_df[available_rating_cols].min(axis=1)
            combined_df['Rating_Std'] = combined_df['Rating_Std'].fillna(combined_df['Rating_Std'].median())
            combined_df['Rating_Range'] = combined_df['Rating_Range'].fillna(combined_df['Rating_Range'].median())


        # Entertainment and ambience score
        entertainment_cols = ['Live Music Rating', 'Comedy Gigs Rating', 'Ambience', 'Lively']
        available_entertainment_cols = [col for col in entertainment_cols if col in combined_df.columns]
        if available_entertainment_cols:
            combined_df['Entertainment_Score'] = combined_df[available_entertainment_cols].mean(axis=1).fillna(combined_df[available_entertainment_cols].median(axis=1))
            combined_df['Entertainment_Score'] = combined_df['Entertainment_Score'].fillna(combined_df['Entertainment_Score'].median())


        # Infrastructure and convenience score
        infra_cols = ['Fire Audit', 'Liquor License Obtained', 'Situated in a Multi Complex',
                      'Dedicated Parking', 'Open Sitting Available']
        available_infra_cols = [col for col in infra_cols if col in combined_df.columns]
        if available_infra_cols:
            combined_df['Infrastructure_Score'] = combined_df[available_infra_cols].sum(axis=1)
            combined_df['Infrastructure_Score'] = combined_df['Infrastructure_Score'].fillna(combined_df['Infrastructure_Score'].median())


        # City and location features
        if 'City' in combined_df.columns:
            city_counts = combined_df['City'].value_counts()
            combined_df['City_Frequency'] = combined_df['City'].map(city_counts).fillna(0)


        # Restaurant age feature (more sophisticated)
        if 'Days_Since_Opening' in combined_df.columns:
            combined_df['Restaurant_Age_Years'] = combined_df['Days_Since_Opening'] / 365.25
            combined_df['Is_New_Restaurant'] = (combined_df['Restaurant_Age_Years'] < 1).astype(int)
            combined_df['Is_Established'] = (combined_df['Restaurant_Age_Years'] > 5).astype(int)
            combined_df['Restaurant_Age_Years'] = combined_df['Restaurant_Age_Years'].fillna(combined_df['Restaurant_Age_Years'].median())


        # Value perception ratio
        if 'Value for Money' in combined_df.columns and 'Food Rating' in combined_df.columns:
            combined_df['Value_Quality_Ratio'] = combined_df['Value for Money'] / (combined_df['Food Rating'].replace(0, 1) + 1)
            combined_df['Value_Quality_Ratio'] = combined_df['Value_Quality_Ratio'].fillna(combined_df['Value_Quality_Ratio'].median())


        # *** Interaction Terms ***
        if 'Social_Media_Score' in combined_df.columns and 'Restaurant Location' in combined_df.columns and pd.api.types.is_numeric_dtype(combined_df['Restaurant Location']):
             print("Creating SocialMedia_x_Location interaction...")
             combined_df['SocialMedia_x_Location'] = combined_df['Social_Media_Score'] * combined_df['Restaurant Location']
        if 'Service_Quality_Score' in combined_df.columns and 'Restaurant Type' in combined_df.columns and pd.api.types.is_numeric_dtype(combined_df['Restaurant Type']):
             print("Creating ServiceQuality_x_RestaurantType interaction...")
             combined_df['ServiceQuality_x_RestaurantType'] = combined_df['Service_Quality_Score'] * combined_df['Restaurant Type']
        if 'Infrastructure_Score' in combined_df.columns and 'Restaurant City Tier' in combined_df.columns and pd.api.types.is_numeric_dtype(combined_df['Restaurant City Tier']):
             print("Creating Infrastructure_x_CityTier interaction...")
             combined_df['Infrastructure_x_CityTier'] = combined_df['Infrastructure_Score'] * combined_df['Restaurant City Tier']
        if 'Restaurant_Age_Years' in combined_df.columns and 'Resturant Tier' in combined_df.columns and pd.api.types.is_numeric_dtype(combined_df['Resturant Tier']):
             print("Creating Age_x_Tier interaction...")
             combined_df['Age_x_Tier'] = combined_df['Restaurant_Age_Years'] * combined_df['Resturant Tier']


        # *** Polynomial Features ***
        print("Creating polynomial features...")
        polynomial_candidates = [
            'Facebook Popularity Quotient', 'Instagram Popularity Quotient',
            'Restaurant Zomato Rating', 'Value for Money', 'Food Rating',
            'Overall Restaurant Rating', 'Social_Media_Score', 'Service_Quality_Score',
            'Entertainment_Score', 'Restaurant_Age_Years', 'Restaurant City Tier', 'Resturant Tier'
        ]
        polynomial_candidates = [col for col in combined_df.columns if col in polynomial_candidates and pd.api.types.is_numeric_dtype(combined_df[col])]


        if polynomial_candidates:
            combined_df_poly = combined_df[polynomial_candidates].copy()
            # No need to fillna here if median imputation was done earlier

            # Fit on the combined data to ensure consistent transformation
            self.poly.fit(combined_df_poly)
            poly_features = self.poly.transform(combined_df_poly)

            # Create a DataFrame for polynomial features with meaningful names
            poly_feature_names = self.poly.get_feature_names_out(polynomial_candidates)
            poly_df = pd.DataFrame(poly_features, columns=poly_feature_names, index=combined_df.index)

            # Drop original polynomial candidate columns from combined_df to avoid duplication
            combined_df = combined_df.drop(columns=polynomial_candidates)

            # Concatenate the new polynomial features
            combined_df = pd.concat([combined_df, poly_df], axis=1)
            print(f"Added {poly_df.shape[1]} polynomial features.")


        # *** More Advanced Encoding (Target Encoding for high cardinality, One-Hot for others) ***
        print("Applying advanced encoding...")

        # Identify categorical columns again
        categorical_cols_final = combined_df.select_dtypes(include=['object']).columns

        # Target Encode high cardinality columns on the training data first to avoid leakage
        high_cardinality_cols = []
        low_cardinality_cols = []
        for col in categorical_cols_final:
            if col not in ['Registration Number'] and col in combined_df.columns:
                unique_count = combined_df[col].nunique()
                if unique_count > 50 and self.train_df.shape[0] > 0:
                    high_cardinality_cols.append(col)
                elif col != 'Registration Number':
                    low_cardinality_cols.append(col)


        # Apply Target Encoding (fit on training data, transform on combined)
        if high_cardinality_cols and self.train_df.shape[0] > 0:
            print(f"Applying Target Encoding to: {high_cardinality_cols}")
            train_combined_part = combined_df[:len(self.train_df)].copy()

            for col in high_cardinality_cols:
                encoder = TargetEncoder(cols=[col])
                # Fit encoder on training data using the categorical column from train_combined_part and the original target
                encoder.fit(train_combined_part, original_train_target) # Use original target for TE
                self.target_encoders[col] = encoder

                # Transform the entire combined_df
                combined_df[f'{col}_TargetEncoded'] = encoder.transform(combined_df)[col]


            # Drop original high cardinality columns
            combined_df = combined_df.drop(columns=high_cardinality_cols)


        # Apply One-Hot Encoding for low cardinality columns
        if low_cardinality_cols:
            print(f"Applying One-Hot Encoding to: {low_cardinality_cols}")
            for col in low_cardinality_cols:
                combined_df[col] = combined_df[col].astype(str)

            combined_df = pd.get_dummies(combined_df, columns=low_cardinality_cols, drop_first=True)


        # Ensure all columns are numeric after encoding, fill any remaining NaNs
        for col in combined_df.columns:
            if combined_df[col].dtype == 'object':
                try:
                    combined_df[col] = pd.to_numeric(combined_df[col])
                except ValueError:
                    print(f"Warning: Column {col} remains object type after encoding. Attempting to drop.")
                    combined_df = combined_df.drop(columns=[col])


        combined_df = combined_df.fillna(combined_df.median()) # Final fillna


        # Split back into train and test
        train_size = len(self.train_df)
        self.processed_train = combined_df[:train_size].copy()
        self.processed_test = combined_df[train_size:].copy()

        # Add the original uncapped target back to the processed train df for model training
        self.processed_train['Annual Turnover'] = original_train_target


        print(f"Processed training data shape: {self.processed_train.shape}")
        print(f"Processed test data shape: {self.processed_test.shape}")

        return self.processed_train, self.processed_test


    def prepare_data(self):
        """Prepare data for models"""
        print(f"\n=== PREPARING DATA FOR MODELS ===")

        # Prepare features (exclude Registration Number and Annual Turnover from features)
        feature_cols = [col for col in self.processed_train.columns
                       if col not in ['Registration Number', 'Annual Turnover']]

        X_train = self.processed_train[feature_cols].values
        # Use the ORIGINAL uncapped target for training tree models
        y_train = self.processed_train['Annual Turnover'].values
        X_test = self.processed_test[feature_cols].values

        print(f"Feature matrix shape: {X_train.shape}")
        print(f"Target shape: {y_train.shape}")

        # Apply log transformation to target (for scaling and potential use in models)
        # Note: Scaling is primarily for models sensitive to scale (like Ridge, though not currently used).
        # Tree models (RF, XGBoost) don't strictly need scaled features or target.
        # We will train the tree models on the original (uncapped) y.
        # y_train_log = np.log1p(y_train) # Log transform might still be useful for understanding distribution

        # Scale features using RobustScaler
        X_train_scaled = self.scaler_features.fit_transform(X_train)
        X_test_scaled = self.scaler_features.transform(X_test)

        # Scale target only for potential future use or if scaling was needed for other models
        # For RF/XGBoost training, we use the original y_train
        # y_train_scaled = self.scaler_target.fit_transform(y_train_log.reshape(-1, 1)).flatten()
        # print(f"Target statistics after log+scaling - Min: {y_train_scaled.min():.3f}, Max: {y_train_scaled.max():.3f}")

        # Return original X_train, y_train for tree models
        return X_train_scaled, y_train, X_test_scaled, len(feature_cols), X_train, y_train


    def create_ensemble_model(self, X_train_scaled, y_train_scaled, X_train_original, y_train_original):
        """Create ensemble with Random Forest and XGBoost"""
        print("\n=== CREATING ENSEMBLE MODEL (RF + XGBoost) ===")

        # Random Forest (uses original scale data)
        self.rf_model = RandomForestRegressor(
            n_estimators=300,
            max_depth=20,
            min_samples_split=5,
            min_samples_leaf=3,
            random_state=42,
            n_jobs=-1
        )
        # Train RF on original features and ORIGINAL uncapped target
        self.rf_model.fit(X_train_original, y_train_original)

        # XGBoost Regressor (uses original scale data)
        self.xgb_model = xgb.XGBRegressor(
            objective='reg:squarederror',
            n_estimators=500,
            learning_rate=0.03,
            max_depth=7,
            subsample=0.85,
            colsample_bytree=0.85,
            random_state=42,
            n_jobs=-1
        )
        # Train XGBoost on original features and ORIGINAL uncapped target
        self.xgb_model.fit(X_train_original, y_train_original)

        return self.rf_model, self.xgb_model

    def make_ensemble_predictions(self, X_test_scaled, X_test_original):
        """Make predictions using ensemble (RF + XGBoost)"""
        # Random Forest predictions (uses original scale data)
        rf_pred = self.rf_model.predict(X_test_original)

         # XGBoost predictions (uses original scale data)
        xgb_pred = self.xgb_model.predict(X_test_original)

        # Simple averaging ensemble
        ensemble_pred = (rf_pred + xgb_pred) / 2.0

        # Ensure predictions are non-negative
        ensemble_pred = np.maximum(0, ensemble_pred)
        rf_pred = np.maximum(0, rf_pred)
        xgb_pred = np.maximum(0, xgb_pred)

        return ensemble_pred, rf_pred, xgb_pred

    def evaluate_model(self, X_train_scaled, y_train_original, X_train_original):
        """Evaluate model performance"""
        print(f"\n=== MODEL EVALUATION ===")

        # Use the ORIGINAL, uncapped target for evaluation
        train_actual = y_train_original

        if self.use_ensemble and self.rf_model and self.xgb_model:
            # Ensemble evaluation
            # make_ensemble_predictions returns predictions on the original scale
            train_pred_ensemble, train_pred_rf, train_pred_xgb = self.make_ensemble_predictions(
                X_train_scaled, X_train_original
            )

            # Calculate RMSE using the ORIGINAL (uncapped) target
            rmse_ensemble = np.sqrt(mean_squared_error(train_actual, train_pred_ensemble))
            rmse_rf = np.sqrt(mean_squared_error(train_actual, train_pred_rf))
            rmse_xgb = np.sqrt(mean_squared_error(train_actual, train_pred_xgb))


            print(f"Ensemble (RF + XGBoost) RMSE (on original target): {rmse_ensemble:,.2f}")
            print(f"Random Forest RMSE (on original target): {rmse_rf:,.2f}")
            print(f"XGBoost RMSE (on original target): {rmse_xgb:,.2f}")


            best_rmse = rmse_ensemble
        else:
             print("No ensemble model available.")
             best_rmse = float('inf')


        print(f"Target RMSE: 12,240,429.84")
        print(f"Performance: {'PASSED' if best_rmse < 12240429.84 else 'NEEDS IMPROVEMENT'}")

        return best_rmse

    def create_submission_file(self, predictions, registration_numbers, filename='optimized_submission.csv'):
        """Create submission file"""
        predictions[predictions < 0] = 0
        submission_df = pd.DataFrame({
            'Registration Number': registration_numbers,
            'Annual Turnover': np.round(predictions).astype(int)
        })

        submission_df.to_csv(filename, index=False)
        print(f"\nSubmission file saved as: {filename}")
        print(f"Submission shape: {submission_df.shape}")
        print(f"Prediction statistics:")
        print(submission_df['Annual Turnover'].describe())

        return submission_df

# Re-run main with the updated preprocessing
predictor = main()
```

## Cross-validation

### Subtask:
Implement cross-validation during training to get a more robust estimate of the model's performance and reduce the risk of overfitting.


**Reasoning**:
Implement K-Fold cross-validation to get a more robust estimate of the ensemble model's performance and reduce overfitting.

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
import numpy as np
import pandas as pd
from sklearn.linear_model import Ridge # Using Ridge as a simple meta-model for stacking

# Ensure X_train_original and y_train_original are available
if 'X_train_original' not in locals() or 'y_train_original' not in locals() or predictor is None:
    print("Required data (X_train_original, y_train_original) or predictor object not found. Re-running the main pipeline...")
    predictor = main()
    if predictor is not None:
         # Access the necessary data from the predictor object after the run
        # prepare_data returns: X_train_scaled, y_train, X_test_scaled, n_features, X_train, y_train
        # We need X_train_original (which is X_train), and y_train_original (which is the y_train returned)
        _, y_train_original, _, _, X_train_original, _ = predictor.prepare_data()
    else:
        print("Failed to re-run main pipeline and get required data.")

if 'X_train_original' in locals() and 'y_train_original' in locals() and predictor is not None:
    print("\n=== IMPLEMENTING ENSEMBLE IMPROVEMENTS (WEIGHTING & STACKING) ===")

    # Define the number of folds for stacking
    n_splits_stacking = 5
    kf_stacking = KFold(n_splits=n_splits_stacking, shuffle=True, random_state=42)

    # Create arrays to hold out-of-fold (OOF) predictions
    oof_preds_rf = np.zeros(X_train_original.shape[0])
    oof_preds_xgb = np.zeros(X_train_original.shape[0])

    # Lists to store validation RMSE for each fold during stacking data generation
    stacking_ensemble_rmse_scores = []
    stacking_rf_rmse_scores = []
    stacking_xgb_rmse_scores = []


    print(f"Generating Out-of-Fold predictions for Stacking using {n_splits_stacking}-fold cross-validation...")
    for fold, (train_index, val_index) in enumerate(kf_stacking.split(X_train_original, y_train_original)):
        print(f"\n--- Stacking Fold {fold+1}/{n_splits_stacking} ---")

        # Split data for the current fold
        X_train_fold, X_val_fold = X_train_original[train_index], X_train_original[val_index]
        y_train_fold, y_val_fold = y_train_original[train_index], y_train_original[val_index]

        # Train the tuned Random Forest model on the training split of the current fold
        fold_rf_model = RandomForestRegressor(**predictor.rf_model.get_params())
        fold_rf_model.fit(X_train_fold, y_train_fold)

        # Train the tuned XGBoost model on the training split of the current fold
        fold_xgb_model = xgb.XGBRegressor(**predictor.xgb_model.get_params())
        fold_xgb_model.fit(X_train_fold, y_train_fold)

        # Make predictions on the validation split (out-of-fold)
        val_pred_rf = fold_rf_model.predict(X_val_fold)
        val_pred_xgb = fold_xgb_model.predict(X_val_fold)

        # Store OOF predictions
        oof_preds_rf[val_index] = np.maximum(0, val_pred_rf) # Ensure non-negative
        oof_preds_xgb[val_index] = np.maximum(0, val_pred_xgb) # Ensure non-negative

        # Evaluate individual models on the validation fold (optional, but good check)
        rmse_rf_fold = np.sqrt(mean_squared_error(y_val_fold, oof_preds_rf[val_index]))
        rmse_xgb_fold = np.sqrt(mean_squared_error(y_val_fold, oof_preds_xgb[val_index]))
        print(f"Fold {fold+1} Val RMSE (RF): {rmse_rf_fold:,.2f}")
        print(f"Fold {fold+1} Val RMSE (XGBoost): {rmse_xgb_fold:,.2f}")

        stacking_rf_rmse_scores.append(rmse_rf_fold)
        stacking_xgb_rmse_scores.append(rmse_xgb_fold)


    print("\n--- Training Meta-Model for Stacking ---")
    # Create the meta-features (OOF predictions from base models)
    meta_features_train = np.column_stack((oof_preds_rf, oof_preds_xgb))

    # Train a meta-model on the OOF predictions and the original training target
    meta_model = Ridge(alpha=1.0) # Using Ridge as a simple linear meta-model
    meta_model.fit(meta_features_train, y_train_original)

    print("Meta-model (Ridge) trained on OOF predictions.")


    # --- Update Predictor Class Methods for Ensemble ---

    class RestaurantTurnoverPredictor_EnhancedEnsemble(RestaurantTurnoverPredictor):
        def __init__(self):
            super().__init__()
            self.meta_model = None # To store the trained meta-model
            self.ensemble_method = 'weighted_average' # Default ensemble method
            self.rf_weight = 0.5 # Default weights for weighted average
            self.xgb_weight = 0.5

        def set_ensemble_method(self, method='weighted_average', rf_weight=0.5, xgb_weight=0.5):
            """Set the ensemble method and weights (if weighted average)"""
            if method not in ['weighted_average', 'stacking']:
                 raise ValueError("Ensemble method must be 'weighted_average' or 'stacking'")
            self.ensemble_method = method
            if method == 'weighted_average':
                 self.rf_weight = rf_weight
                 self.xgb_weight = xgb_weight
                 print(f"Ensemble method set to Weighted Average with RF weight={self.rf_weight}, XGBoost weight={self.xgb_weight}")
            elif method == 'stacking':
                 print("Ensemble method set to Stacking.")


        def make_ensemble_predictions(self, X_test_scaled, X_test_original):
            """Make predictions using the chosen ensemble method"""
            # Get base model predictions (always needed)
            rf_pred = self.rf_model.predict(X_test_original)
            xgb_pred = self.xgb_model.predict(X_test_original)

            # Ensure base predictions are non-negative
            rf_pred = np.maximum(0, rf_pred)
            xgb_pred = np.maximum(0, xgb_pred)

            if self.ensemble_method == 'weighted_average':
                # Weighted average predictions
                ensemble_pred = (self.rf_weight * rf_pred + self.xgb_weight * xgb_pred) / (self.rf_weight + self.xgb_weight)
                print(f"Making Weighted Average predictions with weights RF={self.rf_weight}, XGBoost={self.xgb_weight}")
            elif self.ensemble_method == 'stacking':
                if self.meta_model is None:
                    raise RuntimeError("Meta-model not trained for stacking. Call train_meta_model first.")
                # Create meta-features for test data
                meta_features_test = np.column_stack((rf_pred, xgb_pred))
                # Predict using the meta-model
                ensemble_pred = self.meta_model.predict(meta_features_test)
                print("Making Stacking predictions using meta-model.")
            else:
                # Fallback to simple average or raise error
                print("Unknown ensemble method. Falling back to simple average.")
                ensemble_pred = (rf_pred + xgb_pred) / 2.0


            # Ensure final ensemble predictions are non-negative
            ensemble_pred = np.maximum(0, ensemble_pred)

            return ensemble_pred, rf_pred, xgb_pred


        def evaluate_model(self, X_train_scaled, y_train_original, X_train_original):
            """Evaluate model performance using the chosen ensemble method"""
            print(f"\n=== MODEL EVALUATION ({self.ensemble_method} ensemble) ===")

            # Use the ORIGINAL, uncapped target for evaluation
            train_actual = y_train_original

            if self.use_ensemble and self.rf_model and self.xgb_model:
                # Ensemble evaluation - use the make_ensemble_predictions method
                train_pred_ensemble, train_pred_rf, train_pred_xgb = self.make_ensemble_predictions(
                    X_train_scaled, X_train_original
                )

                # Calculate RMSE using the ORIGINAL (uncapped) target
                rmse_ensemble = np.sqrt(mean_squared_error(train_actual, train_pred_ensemble))
                rmse_rf = np.sqrt(mean_squared_error(train_actual, train_pred_rf))
                rmse_xgb = np.sqrt(mean_squared_error(train_actual, train_pred_xgb))


                print(f"Ensemble ({self.ensemble_method}) RMSE (on original target): {rmse_ensemble:,.2f}")
                print(f"Random Forest RMSE (on original target): {rmse_rf:,.2f}")
                print(f"XGBoost RMSE (on original target): {rmse_xgb:,.2f}")


                best_rmse = rmse_ensemble
            else:
                 print("No ensemble model available.")
                 best_rmse = float('inf')


            print(f"Target RMSE: 12,240,429.84")
            print(f"Performance: {'PASSED' if best_rmse < 12240429.84 else 'NEEDS IMPROVEMENT'}")

            return best_rmse

        def train_meta_model(self, oof_preds_rf, oof_preds_xgb, y_train_original):
            """Train the meta-model for stacking"""
            print("\n--- TRAINING META-MODEL ---")
            meta_features_train = np.column_stack((oof_preds_rf, oof_preds_xgb))
            self.meta_model = Ridge(alpha=1.0) # Use a simple Ridge meta-model
            self.meta_model.fit(meta_features_train, y_train_original)
            print("Meta-model (Ridge) trained.")


    # --- Re-run the pipeline with the enhanced predictor class ---

    print("\n=== RE-RUNNING PIPELINE WITH ENHANCED ENSEMBLE ===")

    # Initialize the enhanced predictor
    predictor_enhanced = RestaurantTurnoverPredictor_EnhancedEnsemble()

    # File paths
    train_path = "Train_dataset_(1)_(1)_(2)_(1).csv"
    test_path = "Test_dataset_(1)_(1)_(2)_(1).csv"

    try:
        # Load data
        train_df, test_df = predictor_enhanced.load_data(train_path, test_path)

        # Explore data (optional in this re-run, but keeps pipeline structure)
        # predictor_enhanced.explore_data()

        # Preprocess data with enhanced features and outlier treatment
        processed_train, processed_test = predictor_enhanced.preprocess_data()

        # Prepare data for models (get original and scaled data)
        X_train_scaled, y_train_original_for_models, X_test_scaled, n_features, X_train_original, y_train_original_check = predictor_enhanced.prepare_data()
        # Ensure X_test_original is created correctly from processed_test
        X_test_original = processed_test[[col for col in processed_test.columns if col not in ['Registration Number', 'Annual Turnover']]].values

        # Train base models (tuned RF + XGBoost) on the full training data
        # We use the tuned models from the previous steps, but train them on the full data again
        # This is necessary for making final test predictions.
        print("\n--- Training Tuned Base Models on Full Training Data ---")
        predictor_enhanced.rf_model = RandomForestRegressor(**predictor.rf_model.get_params())
        predictor_enhanced.rf_model.fit(X_train_original, y_train_original_for_models) # Use original y for tree models

        predictor_enhanced.xgb_model = xgb.XGBRegressor(**predictor.xgb_model.get_params())
        predictor_enhanced.xgb_model.fit(X_train_original, y_train_original_for_models) # Use original y for tree models
        print("Tuned base models trained on full training data.")


        # --- Stacking Implementation ---
        # Generate OOF predictions for stacking (re-doing this step within the enhanced pipeline)
        print("\n--- Generating Out-of-Fold predictions for Stacking ---")
        oof_preds_rf_enhanced = np.zeros(X_train_original.shape[0])
        oof_preds_xgb_enhanced = np.zeros(X_train_original.shape[0])

        kf_stacking_enhanced = KFold(n_splits=n_splits_stacking, shuffle=True, random_state=42)

        for fold, (train_index, val_index) in enumerate(kf_stacking_enhanced.split(X_train_original, y_train_original_for_models)):
             X_train_fold, X_val_fold = X_train_original[train_index], X_train_original[val_index]
             y_train_fold, y_val_fold = y_train_original_for_models[train_index], y_train_original_for_models[val_index] # Use original y

             # Train base models on fold training data
             fold_rf_model = RandomForestRegressor(**predictor_enhanced.rf_model.get_params())
             fold_rf_model.fit(X_train_fold, y_train_fold)

             fold_xgb_model = xgb.XGBRegressor(**predictor_enhanced.xgb_model.get_params())
             fold_xgb_model.fit(X_train_fold, y_train_fold)

             # Make predictions on fold validation data
             val_pred_rf = fold_rf_model.predict(X_val_fold)
             val_pred_xgb = fold_xgb_model.predict(X_val_fold)

             # Store OOF predictions
             oof_preds_rf_enhanced[val_index] = np.maximum(0, val_pred_rf)
             oof_preds_xgb_enhanced[val_index] = np.maximum(0, val_pred_xgb)

        # Train the meta-model for stacking
        predictor_enhanced.train_meta_model(oof_preds_rf_enhanced, oof_preds_xgb_enhanced, y_train_original_for_models) # Use original y


        # --- Evaluate Different Ensemble Methods ---

        # Evaluate Simple Averaging Ensemble (already done in previous steps, but re-evaluate with trained models)
        print("\n--- Evaluating Simple Averaging Ensemble ---")
        predictor_enhanced.set_ensemble_method('weighted_average', rf_weight=0.5, xgb_weight=0.5)
        simple_avg_rmse = predictor_enhanced.evaluate_model(X_train_scaled, y_train_original_for_models, X_train_original)

        # Evaluate Weighted Averaging Ensemble (example weights - could be tuned)
        # Let's try weights based on individual model CV performance (lower RMSE = higher weight)
        # From previous CV step, XGBoost had slightly lower mean RMSE than RF.
        # Let's assign weights inversely proportional to their mean CV RMSE (or just try 0.4 RF, 0.6 XGBoost)
        # Using arbitrary weights 0.4 and 0.6 for demonstration
        print("\n--- Evaluating Weighted Averaging Ensemble (Example Weights 0.4 RF, 0.6 XGBoost) ---")
        predictor_enhanced.set_ensemble_method('weighted_average', rf_weight=0.4, xgb_weight=0.6)
        weighted_avg_rmse = predictor_enhanced.evaluate_model(X_train_scaled, y_train_original_for_models, X_train_original)


        # Evaluate Stacking Ensemble
        print("\n--- Evaluating Stacking Ensemble ---")
        predictor_enhanced.set_ensemble_method('stacking')
        stacking_rmse = predictor_enhanced.evaluate_model(X_train_scaled, y_train_original_for_models, X_train_original)


        # --- Make Final Predictions using the Best Ensemble Method ---
        # Choose the best ensemble method based on training RMSE or cross-validation
        # For now, let's choose Stacking if it performed better than simple/weighted average
        best_ensemble_rmse = min(simple_avg_rmse, weighted_avg_rmse, stacking_rmse)

        if stacking_rmse == best_ensemble_rmse:
            print("\nChoosing Stacking ensemble for final predictions.")
            predictor_enhanced.set_ensemble_method('stacking')
        elif weighted_avg_rmse == best_ensemble_rmse:
             print("\nChoosing Weighted Averaging ensemble for final predictions.")
             # Need to set the weights again if this is chosen
             predictor_enhanced.set_ensemble_method('weighted_average', rf_weight=0.4, xgb_weight=0.6) # Using the example weights
        else:
            print("\nChoosing Simple Averaging ensemble for final predictions.")
            predictor_enhanced.set_ensemble_method('weighted_average', rf_weight=0.5, xgb_weight=0.5)


        # Make final test predictions using the chosen best ensemble method
        test_predictions, _, _ = predictor_enhanced.make_ensemble_predictions(X_test_scaled, X_test_original)


        # Create submission file
        submission_df = predictor_enhanced.create_submission_file(
            test_predictions,
            processed_test['Registration Number'].values
        )

        print(f"\n=== PIPELINE COMPLETED SUCCESSFULLY ===")
        print(f"Final Chosen Ensemble Training RMSE: {best_ensemble_rmse:,.2f}")

    except FileNotFoundError:
        print("\nERROR: Data files not found!")
        print("Please ensure the following files are in your working directory:")
        print("- Train_dataset_(1)_(1)_(2)_(1).csv")
    except Exception as e:
        print(f"\nERROR: {str(e)}")
        import traceback
        traceback.print_exc()
```
